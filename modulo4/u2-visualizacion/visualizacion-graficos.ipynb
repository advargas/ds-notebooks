**Gráfico de barras:**

**Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5)
y1 = [34, 56, 12, 89, 67]

width = 0.80

plt.bar(x, y1, width)
```

```python
y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90] #Nueva lista de valores

# las barras no estarán centradas para que se pueda apilar
width = 0.40

plt.bar(x-0.2, y1, width)
plt.bar(x+0.2, y2, width)

# Leyenda
plt.legend(["Datos 1", "Datos 2"])

# Labels en los ejes
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```

Cambiar los colores a las barras:

```python
plt.bar(x-0.2, y1, width, color='darkblue')
plt.bar(x+0.2, y2, width, color='#f66667')
plt.legend(["Datos 1", "Datos 2"])
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```

Cambiar la dirección de las barras:

```python
plt.barh(x-0.2, y1, width, color='darkblue')
plt.barh(x+0.2, y2, width, color='#f66667')
plt.legend(["Datos 1", "Datos 2"])
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```

**Seaborn**

```python
import seaborn as sns
x1 = ["Barra 1", "Barra 2", "Barra 3", "Barra 4", "Barra 5"]
y1 = [12, 16, 2, 45, 20]
sns.barplot(x=x1,y=y1)
```

Cambiar el color de las barras:

```python
sns.barplot(x=x1, y=y1, color='r', palette ='bright', saturation = 0.9, edgecolor='r', linewidth=5)
```

Añadir una leyenda:

```python
sns.barplot(x=x1, y=y1, color='r', palette ='bright', saturation = 0.9, linewidth=5, hue=x1)
```

Para realizar el cambio de gráficas horizontales solo tendremos que cambiar el orden en el barplot.

```python
sns.barplot(x=y1, y=x1, color='r', palette ='bright', saturation = 0.9, edgecolor='r', linewidth=5)
```

**Gráficos de líneas:**

```python
import matplotlib.pyplot as plt
x1 = [2, 6, 12, 25, 67]
y1 = [12, 16, 2, 45, 20]
plt.plot(x1, y1)
```

```python
import seaborn as sns

x1 = [2, 6, 12, 25, 67]
y1 = [12, 16, 2, 45, 20]

sns.lineplot(x=x1, y=y1)
```

Personalización del gráfico:

```python
sns.lineplot(x=x1, y=y1, marker='o', palette='bright', linewidth=1, ms=8, mec='r', mew=2, mfc='y')
```

Añadir más líneas al gráfico:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(np.array([
    [11,1,0,2,0], [3,8,0,1,0], [0,16,3,0,0], [0,0,12,0,0], [0,0,0,13,0], [0,1,0,0,16]
    ]), range(6), range(5))

sns.lineplot(data=df, markers=['o','d','h','p','*'], palette='bright', linewidth=2, ms=15, mec='r', mew=2, mfc='g')
```

**Gráficos circulares o de sectores:**

```python
import matplotlib.pyplot as plt

x1 = [4, 9, 15, 25, 47]
labels = ["Peras","Manzanas","Melocotones","Sandías", "Melones"]
plt.pie(x1, labels=labels, autopct="%0.1f %%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x1 = [4, 9, 15, 25, 47]
colors = sns.color_palette('pastel')
plt.pie(x=x1, colors = colors)
```
