**Cuestión 1:**

Una empresa necesita bobinas de filamento de 1,75 mm para el departamento de impresión 3D. Necesitan medir qué tan cerca está realmente el diámetro del filamento de 1,75 mm. Utiliza una herramienta de calibre y toma muestras del diámetro cinco veces en el carrete: 1,78, 1,75, 1,72, 1,74, 1,77. Calcule la media y la desviación estándar para este conjunto de valores.

**Notas de Solución:**

Calculamos la media y la desviación estándar tanto con NumPy, como con Pandas.
Inicialmente la desviación estándar calculada en NumPy vs Pandas, es diferente, y aunque cualquiera de las 2 formas de cálculo es correcta, para que coincidan, es neceario sobreescribir **ddof=1**, para que use el mismo delta de grados de libertad que es usado en Pandas.

```python
import numpy as np

# Create an array with the samples of diameter
data = np.array([1.72, 1.74, 1.75, 1.77, 1.78])

# Calculate mean
mean = np.mean(data)
# Calculate standard deviation (using unbiased estimator N-1, such as in Pandas)
std = np.std(data, ddof=1)

print("Con NumPy:")
print(mean)
print(std)
print("La media del diámetro del filamento es de " + str(mean) + " mm.")
print("La desviación estándar es de " + "{:.6f}".format(std))
```

```python
import pandas as pd

data = pd.Series([1.72, 1.74, 1.75, 1.77, 1.78])

print("Con Pandas:")
print(data.describe())
```

**Cuestión 2:**

Un fabricante de móviles dice que el smartphone Z-Phone tiene una vida media de consumo de 42 meses con una desviación estándar de 8 meses. Suponiendo una distribución normal, ¿cuál es la probabilidad de que un Z-Phone aleatorio dado dure entre 20 y 30 meses?

```python
from scipy.stats import norm

mean = 42
std_dev = 8

# Probability that a given random Z-Phone will last between 20 and 30 months.

cdf_upper_limit = norm.cdf(30, mean, std_dev)
cdf_lower_limit = norm.cdf(20, mean, std_dev)

prob = cdf_upper_limit - cdf_lower_limit

print(prob)
print("La probabilidad de que un Z-Phone aleatorio dado dure entre 20 y 30 meses es: " +
      "{:.6f}".format(prob) + " equivalente a " + "{:.3f}".format(prob*100) + "%")
```

**Cuestión 3:**

El primer cliente no tiene la certeza de que el filamento para las impresoras 3D no tenga un diámetro promedio de 1,75 mm como se anuncia. Se midió 34 muestras con una herramienta de medición. La media muestral es 1,715588 y la desviación estándar muestral es 0,029252. ¿Cuál es el intervalo de confianza del 99 % para la media de toda la bobina de filamento?

```python
from math import sqrt
from scipy.stats import norm

def critical_z_value(p):
    norm_dist = norm(loc=0.0, scale=1.0)
    left_tail_area = (1.0 - p) / 2.0
    upper_area = 1.0 - ((1.0 - p) / 2.0)
    return norm_dist.ppf(left_tail_area), norm_dist.ppf(upper_area)

def confidence_interval(p, sample_mean, sample_std, n):
    # Sample size must be greater than 30
    lower, upper = critical_z_value(p)
    lower_ci = lower * (sample_std / sqrt(n))
    upper_ci = upper * (sample_std / sqrt(n))
    return sample_mean + lower_ci, sample_mean + upper_ci

interval = confidence_interval(p=.99, sample_mean=1.715588, sample_std=0.029252, n=34)

print(interval)
print("El intervalo de confianza del 99% para la media de toda la bobina de filamento es: " + str(interval))
```

**Cuestión 4:**

El departamento de marketing de una compañía de telefonía ha iniciado una nueva campaña publicitaria y desea saber si afectó a las ventas, que en el pasado promediaron 10345 dolares por día con una desviación estándar de 552 dolares. La nueva campaña publicitaria duró 45 días y promedió 11,641 dolares en ventas. ¿La campaña afectó a las ventas? ¿Por qué o por qué no? (Utilice una prueba de dos colas para una significación más confiable).

**Notas de Solución:**

La hipótesis nula (H0) es que la campaña publicitaria no afectó las ventas, y la hipótesis alternativa (H1) es que la campaña publicitaria sí afectó las ventas.

En la prueba de 2 colas, nuestro umbral de significación, se reparte a ambas colas, es decir el 5% habitual, ahora daremos 2,5% a cada cola. Si las ventas caen en cualquiera de las dos regiones, nuestra prueba tiene éxito y rechazamos la hipótesis nula.

En este caso, los valores para 2.5% en ambas colas son de 9263 y 11427 USD.
La media de ventas con la nueva campaña publicitaria es de 11641 USD, que cae en la región del 2.5% del límite superior (por encima de 11427 USD), por lo que la prueba tiene éxito y rechazamos la hipótesis nula.

En cuanto al valor-p, debido a que estamos contemplando ambas colas, debemos capturar las probabilidad tanto de estar por debajo de 9263 y por encima de 11427 USD.

Cuando sumamos ambas, obtenemos un valor p de 5%. Que es igual .05, por lo que podemos decir que hay evidencia estadística suficiente para concluir que la campaña publicitaria afectó las ventas de la compañía.

```python
from scipy.stats import norm

# Sales mean is 10345 USD and std dev is 552 USD
mean = 10345
std_dev = 552

# 1. Calculate the umbral range for the 2 tails
x1 = norm.ppf(.025, mean, std_dev)
print(x1) # 9263.09988053389

x2 = norm.ppf(.975, mean, std_dev)
print(x2) # 11426.90011946611

# 2. Probability of 9263 or less USD
p1 = norm.cdf(9263, mean, std_dev)

# 3. Probability of 11427 or more USD
p2 = 1.0 -  norm.cdf(11427, mean, std_dev)

# 4. P-value of both tails
p_value = p1 + p2
print(p_value) # 0.04997885329800411

print("Umbral inferior: " + "{:.3f}".format(x1))
print("Umbral superior: " + "{:.3f}".format(x2))
print("Probabilidad de estar por debajo del umbral inferior: " + "{:.3f}".format(p1))
print("Probabilidad de estar por encima del umbral superior: " + "{:.3f}".format(p2))
print("Valor p final: " + "{:.3f}".format(p_value))
```
