```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
titanic = sns.load_dataset('titanic')
titanic.dtypes
```

```python
titanic.info()
```

Parece ser que nuestro dataset contiene 3 columnas con datos perdidos: age (edad), deck (tipo de camarote) y embark_town (ciudad en las que los pasajeros embarcaron).

```python
titanic.isnull().sum()
sns.heatmap(titanic.isnull(), cbar = False).set_title("Missing values heatmap")
```

En este caso vamos a partir por eliminar la variable deck (dado el elevado volumen de casos) y rellenar las variables age y embarked con el valor más frecuente al ser variables categóricas:

```python
titanic.drop(['deck'],axis=1,inplace=True)
titanic['age']=titanic['age'].fillna(titanic['age'].mode()[0])
titanic['embarked']=titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.hist(bins=10,figsize=(9,7),grid=False);
```

```python
sns.boxplot(titanic['fare'])
```

Vamos a obtener a continuación, primero, segundo y tercer cuartil, y el rango intercuartil para la variable tarifa o fare:

```python
Q1 = titanic['fare'].quantile(0.25)
Q2 = titanic['fare'].quantile(0.5)
Q3 = titanic['fare'].quantile(0.75)

print("Q1: " + str(Q1))
print("Q2: " + str(Q2))
print("Q3: " + str(Q3))

IQR = Q3 - Q1
print("IQR: " + str(IQR))
```

Ya hemos presentado brevemente que dentro del tratamiento de valores atípicos una de las técnicas es recortar los valores o trimming. Podemos usar el IQR o rango intercuartil para filtrar aquellos casos cuyo valor supere el umbral, véase a continuación la forma de hacerlo:

```python
whisker_width = 1.5
Fare_outliers = titanic[(titanic['fare'] < Q1 - whisker_width*IQR) | (titanic['fare'] > Q3 + whisker_width*IQR)]
Fare_outliers.head()
```

Por último, veamos un análisis de densidad bivariante en donde enfrentamos las variables de clase y edad. Podemos ver cómo tanto las clases segunda y tercera tienden a ser más jóvenes que la primera.

```python
titanic.age[titanic['class'] == 'First'].plot(kind='kde')    
titanic.age[titanic['class']  == 'Second'].plot(kind='kde')
titanic.age[titanic['class']  == 'Third'].plot(kind='kde')
 # plots an axis lable
plt.xlabel("Age")    
plt.title("Age Distribution within classes")
# sets our legend for our graph.
plt.legend(('1st Class', '2nd Class','3rd Class'),loc='best');
```

Cuando queremos relacionar dos variables categóricas, usamos un tipo especial llamada tabulación cruzada. Las categorías de una variable determinan las filas de la tabla y las categorías de la otra variable determinan las columnas. Cada celda contiene el número de veces que ocurrió una combinación particular entre ambas categorías. 

Para ello, gracias a Pandas podemos hacer una tabulación cruzada rápida mediante el comando pd.crosstab():

```python
pd.crosstab(titanic["sex"], titanic["survived"])
```

Vamos a obtener la matriz de correlaciones en nuestros datos y vamos a generar una visualización con ella a partir de un mapa de calor.

```python
corr=titanic.corr()#["Survived"]
plt.figure(figsize=(10, 10))

sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features');
```

Observa el siguiente gráfico de dispersión de las variables survided y age. Nos hemos valido del parámetro alpha para modificar la transparencia y que sea más fácil visualizar la información.

```python
plt.scatter(titanic.survived, titanic.age, alpha=0.1)   ## here the plot has to be transparent so we need to pic low alpha value
plt.title("Age wrt Survived")#wrt significa “with reference to”
```

Al igual que en el análisis bivariante podemos analizar la relación entre variables categóricas y numéricas mediante diagramas de caja o boxplots. Estos gráficos a menudo son más explicativos que los diagramas de dispersión. 

Véase el siguiente ejemplo que va a comparar las variables survived y age:

```python
ax = sns.boxplot(x="survived", y="age", 
                data=titanic)
ax = sns.stripplot(x="survived", y="age",
                   data=titanic, jitter=True,
                   edgecolor="gray")
plt.title("Survival by Age",fontsize=12);
```
