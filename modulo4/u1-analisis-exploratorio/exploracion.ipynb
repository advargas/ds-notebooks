**Límites:**

Se define un límite matemático como una sucesión de números que nos indica intuitivamente que los términos de dicha sucesión se aproximan a un único punto o valor, pero nunca lo alcanzan.

```python
from sympy import *

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)

print(result)

#Resultado del código ejecutado: 0
```

Curiosamente, cuando calculamos el número de Euler con límites en SymPy (que se muestra en el siguiente código), SymPy lo reconoce inmediatamente como el número de Euler. Podemos llamar a evalf() para que podamos mostrarlo como un número:

```python
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

print(result) # E
print(result.evalf()) # 2.71828182845905
```

A continuación vemos cómo programar el cálculo de la pendiente mediante pasos infinitesimales.

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m


def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2)

#Resultado del código ejecutado: 4.000010000000827
```

Empezamos definiendo la función a partir de las variables x e y y usamos la función diff 
para obtener la derivada parcial, respecto de x e y. Usamos la función plot3d para representar la curva en el espacio.

```python
from sympy import *
from sympy.plotting import plot3d

# Declare x and y to SymPy
x,y = symbols('x y')

# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3

# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f) # prints 6*x**2
print(dy_f) # prints 9*y**2

# plot the function
plot3d(f);

#Resultado del código ejecutado (imagen)
```

Esta es la regla de la cadena, que dice que para una función dada y (con variable de entrada x) compuesta en otra función z (con variable de entrada y), podemos encontrar la derivada de z con respecto a x multiplicando las dos derivadas respectivas:

```python
from sympy import *

x, y = symbols('x y')


# derivative for first function
# need to underscore y to prevent variable clash
_y = x**2 + 1
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# Calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain) # 6*x*(x**2 + 1)**2
print(dz_dx_no_chain) # 6*x*(x**2 + 1)**2
```

Primero vamos a crear un función que aproxime la integral que llamaremos approximate_integral(). En ella se especifica los argumentos a y b, que son el mínimo y el máximo del rango x de mi función f.

Luego tendremos n que es el número de rectángulos a empaquetar y f nuestra función en sí. En este caso vamos integrar la función x^2+1 mediante cinco rectángulos, usando el método de suma de Riemann.

```python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0

    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)

    return total_sum * delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)

print(area)

#Resultado del código ejecutado: 1.33
```

**Matrices:**

**Ejemplo 1:**

En este caso vamos a aplicar una multiplicación de un vector como una matriz.
Primero construímos una matriz, utilizando np.array de la librería NumPy, donde declaramos los valores.
Luego, volvemos a crear otra variable llamada v, donde declaramos el vector.
Finalmente utilizamos la función dot para aplicar la multiplicación entre ambos elementos.



```python
# compose basis matrix with i-hat and j-hat
basis = np.array(
    [[3, 0],
     [0, 2]]
 )

# declare vector v
v = np.array([1,1])

# create new vector
# by transforming v with dot product
new_v = basis.dot(v)

print(new_v) # [3, 2]
```

**Ejemplo 2:**

En este ejemplo, vamos a realizar una transformación. Aquí declaramos la matriz base mediante dos array i_hat y j_hat. Vemos que la matriz puede construirse como en el ejemplo anterior, todo a la vez, o como aquí por separado. Ambas formas son válidas.

Luego vamos aplicar una transformación de transposición donde las filas irán las posiciones de las columnas, es decir un volteo.

Finalmente, aplicamos la multiplicación entre el vector y la matriz.

```python
# Declare i-hat and j-hat
i_hat = np.array([2, 0])
j_hat = np.array([0, 3])

# compose basis matrix using i-hat and j-hat
# also need to transpose rows into columns
basis = np.array([i_hat, j_hat]).transpose()


# declare vector v 0
v = np.array([2,1])

# create new vector
# by transforming v with dot product
new_v = basis.dot(v)

print(new_v) # [4, 3]
```

**Ejemplo 3:**

En este caso vamos a multiplicar matrices. Lo primero de todo es declarar ambas matrices. Antes de nada, vamos a especificar las filas y las columnas mediante los vectores i_hat y j_hat. Luego combinamos ambas matrices mediante @, podemos utilizar matmul() si se prefiere. Tras esto vamos a multiplicar la matriz combinada con un vector.

```python
##Aplicacióm de transformaciones y multiplicación
# Transformation 1
i_hat1 = np.array([0, 1])
j_hat1 = np.array([-1, 0])
transform1 = np.array([i_hat1, j_hat1]).transpose()
print(transform1)

# Transformation 2
i_hat2 = np.array([1, 0])
j_hat2 = np.array([1, 1])
transform2 = np.array([i_hat2, j_hat2]).transpose()
print(transform2)

# Combine Transformations
combined = transform2 @ transform1

# Test
print("COMBINED MATRIX:\n {}".format(combined))

v = np.array([1, 2])
print(combined.dot(v))  # [-1, 1]

##Transformaciones por separado dan el mismo resultado
rotated = transform1.dot(v)
print(rotated)

sheered = transform2.dot(rotated)
print(sheered) # [-1, 1]
```

```python
##Aplicar las mismas transformaciones a la inversa 
# Transformation 1
i_hat1 = np.array([0, 1])
j_hat1 = np.array([-1, 0])
transform1 = np.array([i_hat1, j_hat1]).transpose()

# Transformation 2
i_hat2 = np.array([1, 0])
j_hat2 = np.array([1, 1])
transform2 = np.array([i_hat2, j_hat2]).transpose()

# Combine Transformations, apply sheer first and then rotation
combined = transform1 @ transform2

# Test
print("COMBINED MATRIX:\n {}".format(combined))
```

**Autovalores y autovectores:**

En este caso vamos a calcular los autovalores y autovectores de una matriz. Para ello, vamos a hacer uso de una serie de funciones de la librería NumPy. En este caso para poder calcular ambos elementos, solo necesitamos aplicar la función numpy.linalg.eig sobre la matriz, como se ve en el siguiente ejemplo.

```python
from numpy import array, diag
from numpy.linalg import eig, inv

A = array([
    [1, 2],
    [4, 5]
])

eigenvals, eigenvecs = eig(A)

print("EIGENVALUES")
print(eigenvals)
print("\nEIGENVECTORS")
print(eigenvecs)
```

Imagenimos que queremos deshacer la operación anterior. Podemos realizar la inversa mediante el uso de la función np.linalg.inv. Primero debemos invertir la matriz de autovectores, y calcular la matriz diagonal mediante los autovalores. Finalmente multiplicamos la matriz de autovectores, la diagonal y su inversa, dando lugar a la matriz original

```python
#Reconstrucción de la matriz a partir de los autovalores y autovectores
A = array([
    [1, 2],
    [4, 5]
])

eigenvals, eigenvecs = eig(A)

print("EIGENVALUES")
print(eigenvals)
print("\nEIGENVECTORS")
print(eigenvecs)

print("\nREBUILD MATRIX")
Q = eigenvecs
R = inv(Q)

L = diag(eigenvals)
B = Q @ L @ R

print(B)
```

Mediana:

```python
# Number of pets each person owns
sample = [0, 1, 5, 7, 9, 10, 14]

def median(values):
    ordered = sorted(values)
    print(ordered)
    n = len(ordered)
    mid = int(n / 2) - 1 if n % 2 == 0 else int(n/2)

    if n % 2 == 0:
        return (ordered[mid] + ordered[mid+1]) / 2.0
    else:
        return ordered[mid]

print(median(sample))
```

Moda:

```python
# Number of pets each person owns
from collections import defaultdict

sample = [1, 3, 2, 5, 7, 0, 2, 3]

def mode(values):
    counts = defaultdict(lambda: 0)

    for s in values:
        counts[s] += 1

    max_count = max(counts.values())
    modes = [v for v in set(values) if counts[v] == max_count]
    return modes

print(mode(sample))

#Resultado del código ejecutado: [2, 3]
```

Distribución normal:

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Creating a series of data of in range of 1-50.
x = np.linspace(1,50,200)
 
#Creating a Function.
def normal_dist(x , mean , sd):
    pdf = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return pdf
 
#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)
 
#Apply function to the data.
pdf = normal_dist(x,mean,sd)
 
#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('PDF')

#Resultado del código ejecutado: (imagen)
```

Intervalo de confianza:

```python
from math import sqrt
from scipy.stats import norm

def critical_z_value(p):
    norm_dist = norm(loc=0.0, scale=1.0)
    left_tail_area = (1.0 - p) / 2.0
    upper_area = 1.0 - ((1.0 - p) / 2.0)
    return norm_dist.ppf(left_tail_area), norm_dist.ppf(upper_area)

def confidence_interval(p, sample_mean, sample_std, n):
    # Sample size must be greater than 30
    lower, upper = critical_z_value(p)
    lower_ci = lower * (sample_std / sqrt(n))
    upper_ci = upper * (sample_std / sqrt(n))
    return sample_mean + lower_ci, sample_mean + upper_ci
    
print(confidence_interval(p=.95, sample_mean=64.408, sample_std=2.05, n=31))
# (63.68635915701992, 65.12964084298008)
#Resultado del código ejecutado: (63.68635915701992, 65.12964084298008)
```

Bootstrap:

```python
import pandas as pd 

df = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Casestudy/titanic_train.csv')
results = []

for nrepeat in range(1000):
    sample = df.resample(df['Fare'])
    results.append(sample.median())
results = pd.Series(results)

print('Bootstrap Statistics:')
print('original: ',df['Fare'].median())
print('bias', results.mean() - df['Fare'].median())
print('std. error', results.std())
```
