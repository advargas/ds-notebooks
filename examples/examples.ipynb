### **Ejercicio 1**
Implementación del Método de Montecarlo.

````python
# Ejercicio 1
import random
import time

def monte_carlo_pi (nsamples):
  acc = 0
  for i in range(nsamples):
    x = random.random()
    y = random.random()
    if(x**2 + y**2) < 1.0:
      acc += 1
  return 4.0 * acc/nsamples

start = time.time()

print(monte_carlo_pi(10_000_000))

end = time.time()
print("Tiempo que tarda en ejecutar = %s" % (end - start))

# Resultado del código ejecutado
# 3.142306
````

Con Numba

````python
# Método Montecarlo con Numba

import random
import numba #Importamos numba
import time

@numba.jit() #Usamos su decorador para compilar esta función.
def monte_carlo_pi (nsamples):
  acc = 0
  for i in range(nsamples):
    x = random.random()
    y = random.random()
    if(x**2 + y**2) < 1.0:
      acc += 1
  return 4.0 * acc/nsamples

start = time.time()

print(monte_carlo_pi(10_000_000))

end = time.time()
print("Tiempo que tarda en ejecutar = %s" % (end - start))

# Resultado del código ejecutado
# 3.1412628
````

La persona que vende pasajes aéreos en la compañía de turismo «La Libertad» pierde mucho tiempo calculando el precio de los pasajes. Esto se debe a que el precio del pasaje depende de: una tarifa básica, la temporada, la compañía aérea, la edad del pasajero y si este es estudiante o no.

Usted debe hacer un programa que ayude a calcular el precio de un pasaje Bogotá-Tokio, teniendo en cuenta que:

La compañía «ALAS» incrementa el valor de sus pasajes en un 30% en alta temporada mientras que la compañía «VOLAR» lo incrementa en solo 20%.

Ambas compañías descuentan el 50% si el pasajero es menor de edad; además, la compañía «VOLAR» tiene un recargo de $100000 para los pasajeros mayores de 60 años para cubrir el seguro de vida.

Los estudiantes que viajan por «ALAS» y que no son menores de edad, tienen un descuento del 10% en temporada baja.

La tarifa básica Bogotá-Tokio reglamentaria es de 5 millones de pesos.

Tenga en cuenta que los descuentos y aumentos deben hacerse sobre la tarifa base.

Por ejemplo, un pasajero de la tercera edad en temporada alta para «VOLAR» recibirá el recargo sobre la tarifa base, no sobre la tarifa con el seguro de vida, y un menor de edad que vuele en temporada alta con «ALAS» recibirá un 20% de descuento, que equivale a 30% de aumento menos 50% de descuento.

````python

def calcular_precio_pasaje(temporada:str, compania:str, edad:int, estudiante:bool)->int:

  tarifa_basica = 5000000
  precio_final = tarifa_basica

  if temporada == 'ALTA':
    if compania == 'ALAS':
      precio_final += tarifa_basica * 0.30
    else:
      precio_final += tarifa_basica * 0.20

  if edad < 18:
    precio_final -= tarifa_basica * 0.5
  elif edad > 60 and compania == 'VOLAR':
    precio_final += 100000
  
  if edad > 18:
    if estudiante and compania == 'ALAS' and temporada == 'BAJA':
      precio_final -= tarifa_basica * 0.10

  return precio_final


print(calcular_precio_pasaje("ALTA", "ALAS", 20, True))
print(calcular_precio_pasaje("BAJA", "ALAS", 70, False))
print(calcular_precio_pasaje("ALTA", "VOLAR", 20, True))
print(calcular_precio_pasaje("BAJA", "VOLAR", 70, False))
print(calcular_precio_pasaje("BAJA", "ALAS", 15, True))
````

Picas y Fijas es un juego en el que dos personas intentan adivinar un número de 4 dígitos que mantiene en secreto el otro jugador. En cada turno, un jugador propone un número de 4 dígitos y el otro jugador debe informar la cantidad de picas y la cantidad de fijas de ese número.

Cada pica significa que en el número propuesto hay un dígito que también está en el número secreto, pero en una posición diferente.

Cada fija significa que en el número propuesto hay un número que también está en el número secreto y que está en la misma posición.

Por ejemplo, si el número secreto es 5678 y el número que el otro jugador propone es 6579, la respuesta sería *2 picas y 1 fija” porque 5 y 6 están en las posiciones equivocadas y porque el 7 está en la posición correcta.

El ganador del juego es el jugador que encuentre el número del otro en la menor cantidad de intentos.

En este ejercicio usted debe escribir una función que calcule la cantidad de fijas dados un número secreto entre 1000 y 9999 y un número propuesto también entre 1000 y 9999.

````python
def contar_fijas(numero_secreto:int, numero_propuesto:int)->int:

  digitos_secreto = list(map(int, str(numero_secreto)))
  digitos_propuesto = list(map(int, str(numero_propuesto)))

  fijas = 0

  if digitos_propuesto[0] == digitos_secreto[0]:
    fijas += 1
  if digitos_propuesto[1] == digitos_secreto[1]:
    fijas += 1
  if digitos_propuesto[2] == digitos_secreto[2]:
    fijas += 1
  if digitos_propuesto[3] == digitos_secreto[3]:
    fijas += 1

  return fijas


print(contar_fijas(5678, 6579))
````

Picas y Fijas es un juego en el que dos personas intentan adivinar un número de 4 dígitos que mantiene en secreto el otro jugador. En cada turno, un jugador propone un número de 4 dígitos y el otro jugador debe informar la cantidad de picas y la cantidad de fijas de ese número.

Cada pica significa que en el número propuesto hay un dígito que también está en el número secreto, pero en una posición diferente.

Cada fija significa que en el número propuesto hay un número que también está en el número secreto y que está en la misma posición.

El ganador del juego es el jugador que encuentre el número del otro en la menor cantidad de intentos.

En este ejercicio usted debe escribir una función que calcule la cantidad de picas dados un número secreto entre 1000 y 9999 y un número propuesto también entre 1000 y 9999.

````python
def contar_picas(numero_secreto:int, numero_propuesto:int)->int:

  digitos_secreto = list(map(int, str(numero_secreto)))
  digitos_propuesto = list(map(int, str(numero_propuesto)))

  picas = 0

  if digitos_propuesto[0] == digitos_secreto[1] \
    or digitos_propuesto[0] == digitos_secreto[2] or digitos_propuesto[0] == digitos_secreto[3]:
    picas += 1
  if digitos_propuesto[1] == digitos_secreto[0] \
    or digitos_propuesto[1] == digitos_secreto[2] or digitos_propuesto[1] == digitos_secreto[3]:
    picas += 1
  if digitos_propuesto[2] == digitos_secreto[0] \
    or digitos_propuesto[2] == digitos_secreto[1] or digitos_propuesto[2] == digitos_secreto[3]:
    picas += 1
  if digitos_propuesto[3] == digitos_secreto[0] \
    or digitos_propuesto[3] == digitos_secreto[1] or digitos_propuesto[3] == digitos_secreto[2]:
    picas += 1

  return picas


print(contar_picas(5678, 6579))
````
