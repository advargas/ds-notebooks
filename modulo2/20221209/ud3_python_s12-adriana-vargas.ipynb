### Ejercicio 12
Escribir un programa que lea un entero positivo, n, introducido por el usuario y después muestre en pantalla la suma de todos los enteros desde 1 hasta n. La suma de los n primeros enteros positivos puede ser calculada de la siguiente forma:
\begin{equation}
\mathrm{suma} = \frac{n \left( n + {1} \right)}{{2}}
\end{equation}

````python
# Ejercicio 12
try:
  n = int(input("Introduce un número: "))
  sum = (n * (n + 1)) / 2
  print("La suma de los enteros de 1 a n es : ", sum)

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 13
Escribir un programa que pida al usuario dos números enteros y muestre por pantalla:
```
La división de n entre m da un cociente c y un resto r
```
Donde n y m son los números introducidos, c es el cociente de la división entera y r el resto.

Pista: en Python 3.x.y el operador `/` lleva a cabo la división real, mientras que `//` lleva a cabo la entera, y `%` nos permite encontrar el módulo de un número frente a otro, o lo que es lo mismo, el resto de la división entera de un número entre otro.

````python
# Ejercicio 13
try:
  n = int(input("Introduce el número dividendo: "))
  m = int(input("Introduce el número divisor: "))

  c = n // m
  r = n % m
  print("La división de", n, "entre", m, "da un cociente", c, "y un resto", r)

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 15
Una juguetería tiene mucho éxito en dos de sus productos: payasos y muñecas. Suele hacer venta por correo y la empresa de logística les cobra por peso de cada paquete así que deben calcular el peso de los payasos y muñecas que saldrán en cada paquete a demanda. Cada payaso pesa 112 g y cada muñeca 75 g. Escribir un programa que lea el número de payasos y muñecas vendidos en el último pedido y calcule el peso total del paquete que será enviado.

````python
# Ejercicio 15
try:
  clown_weight = 112
  doll_weight = 75

  clown_number = int(input("Introduce el número de payasos: "))
  doll_number = int(input("Introduce el número de muñecas: "))

  total_weight = (clown_number * clown_weight) + (doll_number * doll_weight)

  print("El peso total del paquete a enviar es de", total_weight, "g")

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 16
Una panadería vende barras de pan a 3.49€ cada una. El pan que no es el día tiene un descuento del 60%. Escribir un programa que comience leyendo el número de barras vendidas que no son del día. Después el programa debe mostrar el precio habitual de una barra de pan, el descuento que se le hace por no ser fresca y el coste final total.

````python
# Ejercicio 16
try:
  bar_cost = 3.49
  bars_of_bread = int(input("Introduce el número de barras de pan que no son del día: "))
  
  total_cost = bar_cost * bars_of_bread
  discount = total_cost * 0.6
  final_cost = total_cost * 0.4

  print("El precio habitual de una barrada de pan es", bar_cost, "€ cada una")
  print("El descuento por no ser del día es", round(bar_cost * 0.6, 2), "€ cada una")
  print("Coste total sin descuento es", round(total_cost, 2))
  print("El descuento total es", round(discount, 2))
  print("Coste final total con descuento es", round(final_cost, 2))

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 17
Escribir un programa que le solicite al usuario ingresar una fecha formada por 8 números, donde los primeros cuatro representan el año, los siguientes dos el mes y los últimos dos el día (AAAAMMDD). Este dato debe guardarse en una variable con tipo int (número entero). Finalmente, mostrar al usuario la fecha con el formato "DD / MM / AAAA".

````python
# Ejercicio 17
from datetime import datetime

try:
  date_int = input("Introduce una fecha en el formato AAAAMMDD: ")
  date = datetime.strptime(str(date_int), '%Y%m%d')
  
  print("La fecha es:", date.strftime("%d/%m/%Y"))
except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 19
Escribir un programa que solicite al usuario el ingreso de una temperatura en escala Fahrenheit (debe permitir decimales) y le muestre el equivalente en grados Celsius redondeado a 2 decimales. 
\begin{equation}
C = \frac{5}{{9}} \left(F - 32 \right)
\end{equation}

Pista: la función `round(n, d)` en Python redondea el número n a un número de dígitos d.

````python
# Ejercicio 19
try:
  f_value = float(input("Introduce la temperatura en escala Fahrenheit: "))
  c_value = (5 * (f_value - 32)) / 9
  
  print("La temperatura en grados Celsius es:", round(c_value, 2))
except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### Ejercicio 20
Escribir un programa que pida al usuario que introduzca una frase en la consola y una vocal en minúscula, y después muestre por pantalla la misma frase pero con la vocal introducida en mayúscula.
```
aprende a programar con Python → Aprende A progrAmAr con Python
```
Pista: utilizar la función `replace(a, b)` que modifica el valor a por el b sobre la cadena de texto, i.e. `"patata".replace("a", "o")` resultaría en `"pototo"`

````python
# Ejercicio 20
import re

sentence = input("Introduce una frase: ")
vowel = input("Introduce una vocal: ")

if re.match("[aeiou]", vowel) and vowel.lower() in sentence.lower():
  print (sentence.replace(vowel, vowel.upper()))
else:
  print ("Vocal inválida o no se encuentra en la frase")
````
