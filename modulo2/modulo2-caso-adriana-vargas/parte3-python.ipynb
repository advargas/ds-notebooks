### **Conversor entre unidades temporales:**


Nombre de función: `conversor_de_tiempo`

Parámetros de entrada:
* num_segundos: Número de segundos a convertir
* unidad: minutos / horas / dias

Salida:
Imprime en pantalla una cadena con el número de segundos ingresados, convertido a la unidad de tiempo indicada:

`{num_segundos}` segundos equivalen a `{valor_convertido}` `{unidad}`.

Alias para unidades:

*   minutos: m, mm, minutes
*   horas: h, hh, hours
*   dias: d, dd, days

```python
# Conversor entre unidades temporales

unidades_validas = {
    "minutos": ("m","mm","minutes"), 
    "horas": ("h","hh","hours"), 
    "dias": ("d","dd","days")
}

def validar_segundos(str_segundos):
  try:
    numero_segundos = float(str_segundos)
    if numero_segundos >= 0:
      return numero_segundos
    else:
      print("Número de segundos debe ser un número positivo")
  except Exception as e:
    print("Número de segundos debe ser un número válido")
  return None

def validar_unidad(unidad):
  for key, aliases in unidades_validas.items():
    if unidad == key or unidad in aliases:
      return key

  print("Unidad debe ser uno de los valores: minutos/horas/dias")
  return None

def conversor_de_tiempo(numero_segundos, unidad):
  resultado = 0
  if unidad == "minutos":
    resultado = round(numero_segundos / 60, 2)
  elif unidad == "horas":
    resultado = round(numero_segundos / (60 * 60), 2)
  else:
    resultado = round(numero_segundos / (60 * 60 * 24), 2)
  return resultado

try:
  print("Conversor entre unidades temporales")

  nueva_conversion = "S"

  while nueva_conversion.upper() == 'S':  
    str_segundos = input("\nIntroduce el número de segundos a convertir: ")
    numero_segundos = validar_segundos(str_segundos)

    if numero_segundos != None:
      unidad = input("Introduce la unidad destino (minutos/horas/dias): ")
      unidad = validar_unidad(unidad)

      if unidad != None:
        resultado = conversor_de_tiempo(numero_segundos, unidad)
        print("%d segundos equivalen a %.2f %s." % (numero_segundos, resultado, unidad))

    nueva_conversion = input("Realizar otra conversión? S/N ")
  
except Exception as e:
  print(e)
  print("Ha ocurrido un error al convertir los datos")
```
