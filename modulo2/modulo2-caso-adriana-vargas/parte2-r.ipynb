### **Conversor entre unidades temporales:**


Nombre de función: `conversor_de_tiempo`

Parámetros de entrada:
* num_segundos: Número de segundos a convertir
* unidad: minutos / horas / dias

Salida:
Imprime en pantalla una cadena con el número de segundos ingresados, convertido a la unidad de tiempo indicada:

`{num_segundos}` segundos equivalen a `{valor_convertido}` `{unidad}`.

Alias para unidades:

*   minutos: m, mm, minutes
*   horas: h, hh, hours
*   dias: d, dd, days



```r
# Conversor entre unidades temporales
install.packages('hash', lib = "/usr/local/lib/R/site-library")
library(hash)

unidades_validas <- hash()
unidades_validas["minutos"] = list("m","mm","minutes") 
unidades_validas["horas"] = list("h","hh","hours") 
unidades_validas["dias"] = list("d","dd","days")

validar_segundos <- function(str_segundos) {
  tryCatch(
    expr = {
      numero_segundos = as.integer(str_segundos)
      if (numero_segundos >= 0) {
        return(numero_segundos)
      } 
    },
    error = function(e) {
    },
    warning = function(w) {
    }
  )
  print("Número de segundos debe ser un número positivo")
  flush.console()
  return(-1)
}

validar_unidad <- function(unidad) {
  for (key in keys(unidades_validas)) {
    if (unidad == key) {
      return(key)
    } else {
      for (value in unidades_validas[[key]]) {
        if (unidad == value) {
          return(key)
        }
      }
    }
  }
  print("Unidad debe ser uno de los valores: minutos/horas/dias")
  flush.console()
  return("")
}

conversor_de_tiempo <- function(numero_segundos, unidad) {
  resultado <- 0
  if (unidad == "minutos") {
    resultado = round(numero_segundos / 60, 2)
  } else if (unidad == "horas") {
    resultado = round(numero_segundos / (60 * 60), 2)
  } else {
    resultado = round(numero_segundos / (60 * 60 * 24), 2)
  }
  return(resultado)
}

print("Conversor entre unidades temporales")

nueva_conversion = "S"

while (toupper(nueva_conversion) == 'S') {  
  str_segundos = readline("\nIntroduce el número de segundos a convertir: ")
  numero_segundos = validar_segundos(str_segundos)

  if (numero_segundos != -1) {
    unidad = readline("Introduce la unidad destino (minutos/horas/dias): ")
    unidad = validar_unidad(tolower(unidad))
    
    if (unidad != "") {
      resultado = conversor_de_tiempo(numero_segundos, unidad)
      cat(sprintf("%d segundos equivalen a %.2f %s.\n", numero_segundos, resultado, unidad))
      flush.console()
    }
  }

  nueva_conversion = readline("Realizar otra conversión? S/N ")
}
```
