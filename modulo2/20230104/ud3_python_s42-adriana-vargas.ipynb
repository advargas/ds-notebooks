## Numpy

### **Ejercicio 1**

Escribir un programa que cree:
1. Un array vacío
2. Un array de ceros
3. Un array de valores 1 a 5
4. Una matriz diagonal de valores 1 a 5

Y nos muestre por cada uno de ellos sus dimensiones y el número de elementos que contiene.

```python
# Ejercicio 1

import numpy as np

def print_arr(arr):
  print("Array:\n", arr)
  print("Dimensiones:", arr.ndim)
  print("Size:", arr.size)

print_arr(np.empty([5]))
print_arr(np.zeros([10]))
print_arr(np.arange(1,6))
print_arr(np.diagflat(np.arange(1,6)))
```

### **Ejercicio 2**
Escribir un programa que, dados dos arrays, lleve a cabo:
1. La unión
2. La intersección
3. La suma
4. La multiplicación por un valor constante


```python
# Ejercicio 2

import numpy as np

x = np.array([7, 5, 89, 22])
y = np.array([22, 7, 78, 34])

print("x=", x)
print("y=", y)
print("Unión=", np.union1d(x, y))
print("Intersección=", np.intersect1d(x, y))
print("Suma=", x + y)
print("Multiplicación * 5=", x * 5)
```

## Ejercicio 3
Escribir un programa que dados dos arrays los compare.

```python
# Ejercicio 3

import numpy as np

x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
  
print("x:", x)
print("y:", y)
print("x == y:", np.array_equal(x, y))
print("x > y:", np.greater(x, y))
print("x >= y:", np.greater_equal(x, y))
print("x < y:", np.less(x, y))
print("x <= y:", np.less_equal(x, y))
```

### **Ejercicio 4**
Escribir un programa que cree un array de temperaturas en grados fahrenheit y transforme todos sus elementos a grados centígrados. Recordad que la fórmula utilizada para la conversión es:
\begin{equation}
C = \frac{5}{{9}} \left(F - 32 \right)
\end{equation}

```python
# Ejercicio 4

def farenheit_to_celsius(f_value):
  return round((5 * (f_value - 32)) / 9, 2)

in_farenheit = [86, 39.2, 57.2, 76.1, 100, 80.6]
to_celsius = np.vectorize(farenheit_to_celsius)
in_celsius = to_celsius(in_farenheit)
print("Farenheit:", in_farenheit)
print("Celsius:", in_celsius)
```

### **Ejercicio 5**
Escribir un programa que cree una matriz de dimensión 3x5 y otra de dimensión 5x4, con valores aleatorios, y calcule su multiplicación.

```python
# Ejercicio 5

a = np.random.random((3,5))
b = np.random.random((5,4))

print("a:\n", a)
print("b:\n", b)
print("Producto punto:\n", np.dot(a, b))
print("Multiplicación:\n", np.matmul(a, b))
```
