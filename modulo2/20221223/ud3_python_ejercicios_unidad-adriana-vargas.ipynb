### **Ejercicio 1**
Solicitar al usuario que introduzca su dirección de email. Imprimir un mensaje indicando si la dirección es válida o no, valiéndose de una función para decidirlo. Una dirección se considerará válida si contiene el símbolo @.

````python
# Ejercicio 1
import re

regex = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')

def is_valid_email(email):
    if re.fullmatch(regex, email):
      print("Dirección de email válida")
    else:
      print("Dirección de email inválida")
try:
  email = input("Introduce una dirección de email:  ")
  is_valid_email(email)
  
except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de dato! ¡¡Adiós!!")
````

### **Ejercicio 4**
Una inmobiliaria de una ciudad maneja una lista de inmuebles como la siguiente (Una lista donde cada elemento es un diccionario):
```
[
  {'año': 2000, 'metros': 100, 'habitaciones': 3, 'garaje': True, 'zona': 'A'}, 
  {'año': 2012, 'metros': 60, 'habitaciones': 2, 'garaje': True, 'zona': 'B'},
  {'año': 1980, 'metros': 120, 'habitaciones': 4, 'garaje': False, 'zona': 'A'},
  {'año': 2005, 'metros': 75, 'habitaciones': 3, 'garaje': True, 'zona': 'B'},
  {'año': 2015, 'metros': 90, 'habitaciones': 2, 'garaje': False, 'zona': 'A'}
]
```
Construye una función que permita hacer búsqueda de inmuebles en función de un presupuesto dado. La función recibirá como entrada la lista de inmuebles y un precio, y devolverá otra lista con los inmuebles cuyo precio sea menor o igual que el dado. Los inmuebles de la lista que se devuelva deben incorporar un nuevo par a cada diccionario con el precio del inmueble, donde el precio de un inmueble se calcula con las siguiente fórmula en función de la zona:

    Zona A: precio = (metros x 1000 + habitaciones x 5000 + garaje x 15000) x (1 - antiguedad / 100) 
    Zona B: precio = (metros x 1000 + habitaciones x 5000 + garaje x 15000) x (1 - antiguedad / 100) x 1.5


````python
# Ejercicio 4
import datetime

def get_property_price(p):
  area = p["metros"]
  rooms = p["habitaciones"]
  garage = 1 if p["garaje"] else 0

  today = datetime.date.today()
  years = int(today.strftime("%Y")) - p["año"]

  price = (area * 1000 + rooms * 5000 + garage * 15000) * (1 - years / 100)

  if p["zona"] == 'B':
    price *= 1.5
  return round(price, 2)

def search_properties(properties, budget):
  result = []
  for p in properties:
    p['precio'] = get_property_price(p)
    
    if p['precio'] <= budget:
      result.append(p)
  return result

try:
  properties = [
    {'año': 2000, 'metros': 100, 'habitaciones': 3, 'garaje': True, 'zona': 'A'}, 
    {'año': 2012, 'metros': 60, 'habitaciones': 2, 'garaje': True, 'zona': 'B'},
    {'año': 1980, 'metros': 120, 'habitaciones': 4, 'garaje': False, 'zona': 'A'},
    {'año': 2005, 'metros': 75, 'habitaciones': 3, 'garaje': True, 'zona': 'B'},
    {'año': 2015, 'metros': 90, 'habitaciones': 2, 'garaje': False, 'zona': 'A'}
  ]

  budget = float(input("Ingresa tu presupuesto: "))

  filtered_properties = search_properties(properties, budget)

  if len(filtered_properties) > 0:
    print("Los siguientes inmuebles están dentro del presupuesto:")
    for p in filtered_properties:
      print(p)
  else:
    print("No se encontraron inmuebles dentro del presupuesto:")
  
except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de dato! ¡¡Adiós!!")
````

### **Ejercicio 6**
Escribir un programa para gestionar un listín telefónico con los nombres y los teléfonos de los clientes de una empresa. 

El programa incorporará las siguientes funciones: Crear el fichero con el listín si no existe, consultar el teléfono de un cliente, añadir el teléfono de un nuevo cliente y eliminar el teléfono de un cliente. El listín debe estar guardado en el fichero de texto listin.txt donde el nombre del cliente y su teléfono deben aparecer separados por comas y cada cliente en una línea distinta.

A continuación puedes crear el fichero con los siguiente texto añadido:
```
Pedro Blanco,987654321
Marta Azul,978563412
Lucas Amarillo,966853214
```


````python
# Ejercicio 6
from os.path import exists

contacts_path = "sample_data/listin.txt"
contacts = []

def read_contacts():
  try:
    file_exists = exists(contacts_path)
    
    if file_exists:
      line_contacts = ''
      with open(contacts_path, 'r') as contacts_file:
        contacts.clear()
        for line in contacts_file:
          contact = line.replace("\n","").split(",")
          contacts.append((contact[0], contact[1]))

  except FileNotFoundError as error:
    print(error)

def update_contacts():
  handler = open(contacts_path, 'w')
  for contact in contacts:
    handler.write(contact[0] + "," + str(contact[1]) + "\n")
  handler.close()

def search_contact(name):
  try:
    read_contacts()

    if len(contacts) > 0:
      for contact in contacts:
        if name.upper() in contact[0].upper():
          return contact
  except FileNotFoundError as error:
    print(error)
  else:
    return None

def add_contact(contact):
  contacts.append(contact)
  update_contacts()

def remove_contact(name):
  contact = search_contact(name)
  if contact != None:
    contacts.remove(contact)
    update_contacts()

def print_contact(contact):
  if contact != None:
    print(contact[0] + "," + contact[1])

def print_contacts():
  if len(contacts) == 0:
    read_contacts()

  if len(contacts) > 0:
    for contact in contacts:
      print(contact)
  else:
    print("No hay contactos registrados")

try:
  option = 1
  print("Listín Telefónico")

  while option != 5:
    option = int(input("Seleccione una opción:\n1 - Consultar un contacto\n2 - Añadir un contacto\n3 - Eliminar un contacto\n4 - Ver todos los contactos\n5 - Salir\nIntroduzca una opción:  "))
    
    if option == 1:
      name = input("Ingrese el nombre a buscar: ")
      contact = search_contact(name)
      print_contact(contact)
      
    elif option == 2:
      name = input("Ingrese el nombre: ")
      phone = int(input("Ingrese el teléfono: "))
      add_contact((name, phone))
      print("Contacto agregado")
      print((name, phone))
      
    elif option == 3:
      name = input("Ingrese el nombre a eliminar: ")
      remove_contact(name)
      print("Contacto eliminado")

    elif option == 4:
      print_contacts()

except FileNotFoundError as error:
  print(error)
````

Ahora bien:
1. Construye una función reciba el fichero de cotizaciones y devuelva un diccionario con los datos del fichero por columnas.

2. Construye una función que reciba el diccionario devuelto por la función anterior y cree un fichero en formato csv con el mínimo, el máximo y la media de cada columna.


````python
# Ejercicio 7
import os
from os.path import exists
import locale

def read_records():
  cotizacion_path = "sample_data/cotizacion.csv"
  records = {}

  try:  
    file_exists = exists(cotizacion_path)
    
    if file_exists:
      lines = ''
      
      with open(cotizacion_path, 'r') as cotization_file:
        i = 0

        records.clear()
        records['Final'] = []
        records['Máximo'] = []
        records['Mínimo'] = []
        records['Volumen'] = []
        records['Efectivo'] = []
        
        for line in cotization_file:
          if i > 0:
            line = line.replace("\n","").replace(".","").replace(",",".").split(";")
            records['Final'].append(float(line[1]))
            records['Máximo'].append(float(line[2]))
            records['Mínimo'].append(float(line[3]))
            records['Volumen'].append(float(line[4]))
            records['Efectivo'].append(float(line[5]))
          i += 1
        
  except FileNotFoundError as error:
    print(error)

  return records

def write_results(records):
  results_path = "sample_data/results.csv"
  handler = open(results_path, 'w')
  handler.write("Columna;Mínimo;Máximo;Media\n")

  for key in records.keys():
    numbers = records[key]
    line = key + ";"
    line += locale.format_string('%.2f', min(numbers), grouping=True) + ";"
    line += locale.format_string('%.2f', max(numbers), grouping=True) + ";"
    line += locale.format_string('%.2f', sum(numbers)/len(numbers), grouping=True) + ";"
    print(line)
    handler.write(line + "\n")
  
  handler.close()

try:
  records = read_records()
  print(records)
  write_results(records)

except FileNotFoundError as error:
  print(error)
````

### **Ejercicio 8**
Crea una clase llamada Cuenta que tendrá los siguientes atributos:

- **titular** (que es una persona). Es obligatorio.
- **cantidad** (puede tener decimales). Es opcional.

Construye los siguientes métodos para la clase:

1. Un constructor, donde los datos pueden estar vacíos.
2. Los setters y getters para cada uno de los atributos. El atributo no se puede modificar directamente, sólo ingresando o retirando dinero.
3. `mostrar()`: Muestra los datos de la cuenta.
4. `ingresar(cantidad)`: se ingresa una cantidad a la cuenta, si la cantidad introducida es negativa, no se hará nada.
5. `retirar(cantidad)`: se retira una cantidad a la cuenta. La cuenta puede estar en números rojos.



````python
# Ejercicio 8

class Cuenta:
  
  def __init__(self, titular, cantidad=0):
    self._titular = titular
    self._cantidad = cantidad

  def get_titular(self):
    return self._titular

  def set_titular(self, titular):
    self._titular = titular

  def get_cantidad(self):
    return self._cantidad

  def ingresar(self, cantidad):
    if cantidad > 0:
      self._cantidad += cantidad

  def retirar(self, cantidad):
    if cantidad > 0:
      self._cantidad -= cantidad

  def mostrar(self):
    print("Cuenta")
    print("Titular:", self._titular)
    print("Cantidad:", self._cantidad)
  
print("------------------------")
cuenta1 = Cuenta("Ana", 1000)
cuenta1.ingresar(400)
cuenta1.mostrar()
cuenta1.ingresar(200)
cuenta1.mostrar()
cuenta1.retirar(350)
cuenta1.mostrar()

print("------------------------")
cuenta2 = Cuenta("Pedro", 800)
cuenta2.ingresar(700)
cuenta2.mostrar()
cuenta2.ingresar(400)
cuenta2.mostrar()
cuenta2.retirar(300)
cuenta2.mostrar()
````

### **Ejercicio 9**
Vamos a definir ahora una “Cuenta Joven”, para ello vamos a crear una nueva clase CuentaJoven que deriva de la anterior. Cuando se crea esta nueva clase, además del titular y la cantidad, se debe guardar una bonificación que estará expresada en tanto por ciento.

Construye los siguientes métodos para la clase:

1. Un constructor.
2. Los setters y getters para el nuevo atributo.
3. En esta ocasión los titulares de este tipo de cuenta tienen que ser mayor de edad, por lo tanto hay que crear un método `esTitularValido()` que devuelve `True` si el titular es mayor de edad pero menor de 25 años y false en caso contrario.
4. Además la retirada de dinero sólo se podrá hacer si el titular es válido.
5. El método `mostrar()` debe devolver el mensaje de `“Cuenta Joven”` y la bonificación de la cuenta.

Piensa los métodos heredados de la clase madre que hay que reescribir.


````python
# Ejercicio 9

class Cuenta:
  
  def __init__(self, titular, cantidad=0):
    self._titular = titular
    self._cantidad = cantidad

  def get_titular(self):
    return self._titular

  def set_titular(self, titular):
    self._titular = titular

  def get_cantidad(self):
    return self._cantidad

  def ingresar(self, cantidad):
    if cantidad > 0:
      self._cantidad += cantidad

  def retirar(self, cantidad):
    if cantidad > 0:
      self._cantidad -= cantidad

  def mostrar(self):
    print("Cuenta")
    print("Titular:", self._titular)
    print("Cantidad:", self._cantidad)

class CuentaJoven(Cuenta):
  
  def __init__(self, titular, edad, cantidad, bonificacion):
    super().__init__(titular, cantidad)
    self._edad = edad
    self._bonificacion = bonificacion

  def get_bonificacion(self):
    return self._bonificacion

  def get_edad(self):
    return self._edad

  def esTitularValido(self):
    if self._edad >= 18 and self._edad < 25:
      return True
    return False

  def retirar(self, cantidad):
    if self.esTitularValido():
      super().retirar(cantidad)
    else:
      print("Titular inválido")

  def mostrar(self):
    print("Cuenta Jóven")
    print("Titular:", self._titular)
    print("Edad:", self._edad)
    print("Cantidad:", self._cantidad)
    print("Bonificación: " + str(round(self._bonificacion,2)) + "%")

print("------------------------")
cuenta1 = CuentaJoven("Ana", 22, 1000, 3.5)
cuenta1.ingresar(400)
cuenta1.mostrar()
cuenta1.ingresar(200)
cuenta1.mostrar()
cuenta1.retirar(350)
cuenta1.mostrar()

print("------------------------")
cuenta2 = CuentaJoven("Pedro", 23, 800, 3.2)
cuenta2.ingresar(700)
cuenta2.mostrar()
cuenta2.ingresar(400)
cuenta2.mostrar()
cuenta2.retirar(300)
cuenta2.mostrar()

print("------------------------")
cuenta3 = CuentaJoven("Pepe", 17, 400, 3.2)
cuenta3.ingresar(200)
cuenta3.mostrar()
cuenta3.ingresar(200)
cuenta3.mostrar()
cuenta3.retirar(300)
cuenta3.mostrar()
````
