### **Ejercicio 3**
Crear un archivo llamado personas que contendrá una lista de personas con los siguientes campos:

Leer cada línea y colocar cada empleado en un diccionario dentro de una lista. Mostrar los datos de esta. Datos dentro del fichero:
```
1;Carlos;Pérez;05/01/1989
2;Manuel;Heredia;26/12/1973
```

````python
# Ejercicio 3
from os.path import exists

file_path = "sample_data/personas.txt"

def write_file():
  handler = open(file_path, 'w')
  handler.write("1;Carlos;Pérez;05/01/1989\n")
  handler.write("2;Manuel;Heredia;26/12/1973\n")
  handler.write("3;Rosa;Campos;12/06/1961\n")
  handler.write("4;David;García;25/07/2006\n")
  handler.close()

def read_file():
  try:
    file_exists = exists(file_path)
    people = {}
    
    if file_exists:
      with open(file_path, 'r') as personas_file:
        for line in personas_file:
          person = line.replace("\n","").split(";")
          data = []
          i = 0
          for element in person:
            if i > 0:
              data.append(element)
            i += 1
          people[int(person[0])] = data

    return people
        
  except FileNotFoundError as error:
    print(error)

try:
  print("Lista de Personas")
  write_file()
  people = read_file()
  print(people)

except FileNotFoundError as error:
  print(error)
````

### **Ejercicio 4**
Diseña un programa que obtenga los 100 primeros números primos y los almacene en un fichero de texto llamado primos.txt.


````python
# Ejercicio 4

file_path = "sample_data/primos.txt"

def prime_numbers(x):
  cont = 0
  prime_numbers = []
  for i in range(2, x + 1):
    prime = True
    for j in range(2,i):
      if i == j:
        break
      elif i%j == 0:
        prime = False
      else:
        continue
    if prime == True:
        prime_numbers.append(i)
        cont += 1
        
  return prime_numbers

def update_file(numbers):
  handler = open(file_path, 'w')
  for number in numbers:
    handler.write(str(number) + ",")
  handler.close()

try:
  print("100 Primeros Números Primos")
  numbers = prime_numbers(100)
  update_file(numbers)
  print(numbers)
  print("Fichero actualizado ...")

except FileNotFoundError as error:
  print(error)
````

### **Ejercicio 5**
Escribir una función que pida dos números n y m entre 1 y 10, lea el fichero tabla-n.txt con la tabla de multiplicar de ese número, y muestre por pantalla la línea m del fichero.

````python
# Ejercicio 5
base_path = "sample_data/tablas.txt/"

def read_table(n, m):
  filename = base_path + 'tabla-' + str(n) + '.txt'

  try:
    with open(filename, 'r') as table_file:
      for line in table_file:
        if "x " + str(m) in line:
          return line
  
  except FileNotFoundError as error:
    print(error)
  else:
    return ""

try:
  n = int(input("Ingresa un número n, entre 1 y 10: "))
  m = int(input("Ingresa un número m, entre 1 y 10: "))

  if (n >= 1 and n <= 10) and (m >= 1 and m <= 10):
    table_line = read_table(n, m)
    print(table_line)

except FileNotFoundError as error:
  print(error)
````

### **Ejercicio 6**
Escribir un programa que guarde en un fichero el ticket de compra de una venta de zapatillas. El programa deberá comenzar creando el fichero llamado “ticket” y mostrando el listado de zapatillas de la siguiente forma (con el número por delante):
1. Nike Air Max 90
2. Nike Air Force 1
3. Nike Blazer Mid 77 
4. ...

Le preguntaremos al usuario en bucle que vaya seleccionando las zapatillas que quiere comprar hasta que escriba `*`. Estas seleccionadas tendrán que guardarse en una lista que se llame “carrito”. Cuando escriba `*` tendremos que escribir las zapatillas de la lista carrito en el fichero ticket y en la ultima línea del fichero el precio total que tiene que pagar en el fichero.
```
productos = ("Nike Air Max 90", "Nike Air Force 1", "Nike Blazer Mid 77", "Air Jordan 1 Mid", "Nike SB Shane", "Nike Waffle One SE", "Nike Air Presto", "Nike React Vision")
precios = (139.99, 99.99, 109.99, 129.99, 74.99, 109.89, 85.99, 119.80) 
stocks = (43, 1, 0, 32, 10`, 12, 50, 6)
```

````python
# Ejercicio 6

file_path = "sample_data/ticket.txt"

products = ["Nike Air Max 90", "Nike Air Force 1", "Nike Blazer Mid 77", "Air Jordan 1 Mid", "Nike SB Shane", "Nike Waffle One SE", "Nike Air Presto", "Nike React Vision"]
prices = [139.99, 99.99, 109.99, 129.99, 74.99, 109.89, 85.99, 119.80]
stocks = [43, 1, 0, 32, 10, 12, 50, 6]

def update_file(cart, total):
  handler = open(file_path, 'w')

  handler.write("Producto,Precio,Cantidad\n")
  for product in cart.values():
    line = ""
    for element in product:
      line += str(element) + ", "
    handler.write(line[:len(line)] + "\n")

  handler.write("Total = " + str(round(total,2)))

  handler.close()

def build_menu():
  number = 1
  menu = "Productos Disponibles:\n"
  for product in products:
    menu += str(number) + ". " + product + " " + str(prices[number-1]) + "\n"
    number += 1
  menu += "* - Salir\nAgregar un producto:  "
  return menu

def select_products():
  cart = {}
  total = 0
  
  choice = 1
  menu = build_menu()
    
  while choice != "*":
    choice = input(menu)
    
    if choice != "*":
      index = int(choice) - 1
      if index >= 0 and index < len(products):
        if stocks[index] >= 1:
          if index in cart:
            cart[index][2] += 1
          else:
            cart[index] = [products[index], prices[index], 1]
          total += prices[index]
          print("Producto " + products[index] + " agregado al carrito")
          print("Valor total: " + str(round(total, 2)))
        else:
          print("No hay existencias de este producto")
    else:
      print(str(len(cart)) + " productos seleccionados.")
      print("Valor total: " + str(round(total, 2)))

  return (cart, total)

try:
  print("Carrito de Compras")
  (cart, total) = select_products()
  update_file(cart, total)

except FileNotFoundError as error:
  print(error)
````

### **Ejercicio 8**
Implementar un llavero de contraseñas con ficheros. Para ello nos dan un fichero, una columna donde se encuentran las aplicaciones y las contraseñas.
Tendremos que hacer:
1. Agregar nuevas contraseñas. Debes tener en cuenta que la aplicación y la contraseña que se agregue NO puede existir en el fichero. Si existe se tendrá que solicitar otra contraseña.
2. Buscar una aplicación introducida por teclado y devolver la contraseña correspondiente.
3. Deshacer la última contraseña agregada.
4. Modificar una contraseña que se ha introducido por teclado.
5. Después de realizar cada opción se preguntará si se quiere realizar otra acción. Si selecciona “S” se volverá́ a enseñar el menú́, si selecciona “N” parará el programa.

Datos dentro del fichero:
```
Gmail 20051206Correo! 
Tinder qwerasdf1234# 
Twitter Tw2022+LMP 
Tiktok Tik2022*VTP
Instagram Ins2022!MFR
```

````python
# Ejercicio 8
from os.path import exists

llavero_path = "sample_data/llavero.txt"
passwords = []
new_passwords = []

def read_passwords():
  try:
    file_exists = exists(llavero_path)
    
    if file_exists:
      line_contacts = ''
      with open(llavero_path, 'r') as passwords_file:
        passwords.clear()
        for line in passwords_file:
          password = line.replace("\n","").split(" ")
          if len(password) > 0:
            passwords.append([password[0], password[1]])

  except FileNotFoundError as error:
    print(error)

def update_passwords():
  handler = open(llavero_path, 'w')
  all_passwords = passwords + new_passwords
  for password in all_passwords:
    handler.write(password[0] + " " + str(password[1]) + "\n")
  handler.close()

def search_password(app):
  try:
    read_passwords()

    if len(passwords) > 0 or len(new_passwords) > 0:
      all_passwords = passwords + new_passwords
      for password in all_passwords:
        if app.upper() == password[0].upper():
          return password
  
  except FileNotFoundError as error:
    print(error)
  else:
    return None

def add_password(password):
  try:
    old_password = search_password(password[1])
    if old_password == None:
      new_passwords.append(password)
      return True
    else:
      return False
  
  except FileNotFoundError as error:
    print(error)
  else:
    return None

def modify_password(password):
  try:
    old_password = search_password(password[0])
    if old_password != None:
      all_passwords = passwords + new_passwords
      for existing_password in all_passwords:
        if password[0].upper() == existing_password[0].upper():
          existing_password[1] = password[1]
      return True
    else:
      return False
  
  except FileNotFoundError as error:
    print(error)
  else:
    return None

def remove_password():
  if len(new_passwords) > 0:
    deleted_password = new_passwords[-1]
    new_passwords.remove(deleted_password)
    return deleted_password

  return None

def print_password(password):
  if password != None:
    print(password[0] + " " + password[1])

try:
  option = 1
  print("Llavero")

  while option != 5:
    option = int(input("Seleccione una opción:\n1 - Añadir una contraseña\n2 - Buscar una contraseña\n3 - Eliminar última contraseña agregada\n4 - Modificar una contraseña\n5 - Salir\nIntroduzca una opción:  "))
    
    if option == 1:
      app = input("Ingrese la aplicación: ")
      password = input("Ingrese la contraseña: ")
      added = add_password([app, password])
      if added:
        print("Contraseña agregada")
        print((app, password))
      else:
        print("Contraseña existente, por favor modificar")
      
    elif option == 2:
      app = input("Ingrese la aplicación: ")
      password = search_password(app)
      print_password(password)
      
    elif option == 3:
      password = remove_password()
      if password != None:
        print("Password para aplicación " + password[0] + " eliminada")
      else:
        print("No hay contraseñas agregadas recientemente")

    elif option == 4:
      app = input("Ingrese la aplicación: ")
      password = input("Ingrese la contraseña: ")
      updated = modify_password([app, password])
      if updated:
        print("Contraseña actualizada")
        print((app, password))
      else:
        print("Contraseña no encontrada")

    elif option == 5:
      update_passwords()

except FileNotFoundError as error:
  print(error)
````

## **Ejercicio 9**
En este ejercicio deberemos realizar una función que realice varias tareas sobre un fichero llamado **contador.txt** que almacenará un contador de visitas (será un número):
- Nuestro programa trabajará sobre el fichero contador.txt. Si el fichero está vacío le introduciremos el contador a 0. Si existe simplemente leeremos el valor del contador.
- Luego a partir de un argumento:
  * Si se envía el argumento inc, se incrementará el contador en uno y se mostrará por pantalla.
  * Si se envía el argumento dec, se decrementará el contador en uno y se mostrará por pantalla.
  * Si no se envía ningún argumento (o algo que no sea inc o dec), se mostrará el valor del contador por pantalla.
- Finalmente guardará de nuevo el valor del contador de nuevo en el fichero

````python
# Ejercicio 9
from os.path import exists

contador_path = "sample_data/contador.txt"

def read_counter():
  try:
    file_exists = exists(contador_path)
    counter = 0
    
    if file_exists:
      with open(contador_path, 'r') as counter_file:
        for line in counter_file:
          counter = int(line.replace("\n",""))

    return counter

  except FileNotFoundError as error:
    print(error)

def update_counter(counter):
  handler = open(contador_path, 'w')
  handler.write(str(counter) + "\n")
  handler.close()

def increase():
  counter = read_counter()
  counter += 1
  update_counter(counter)
  return counter

def decrease():
  counter = read_counter()
  if counter > 0:
    counter -= 1
    update_counter(counter)
  return counter
  
try:
  option = "inc"
  print("Contador de Visitas")

  while option == 'inc' or option == 'dec':
    option = input("Seleccione una opción:\ninc - para incrementar el contador\ndec - para decrementar el contador\nIntroduzca una opción:  ")
    
    if option == 'inc':
      counter = increase()
      print("Nuevo contador de visitas:", counter)
      
    elif option == 'dec':
      counter = decrease()
      if counter >= 0:
        print("Nuevo contador de visitas:", counter)
      else:
        print("Contador de visitas en cero")     
      
    else:
      counter = read_counter()
      print("Contador de visitas:", counter)

except FileNotFoundError as error:
  print(error)
````
