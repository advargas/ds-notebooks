### **Ejercicio 2**
Definir dos funciones:
1. Una función llamada area_esfera que, a partir del radio de un esfera, devuelva el valor de su área.
\begin{equation}
\mathrm{A}(r) = {4} {\pi} r^{{2}}
\end{equation}

2. Una función llamada volumen_esfera que, a partir del radio de una esfera, devuelva el valor de su volúmen.
\begin{equation}
\mathrm{V}(r) = \frac{{4}}{{3}} {\pi} r^{{3}}

Bastaría con definir en una variable una estimación del numero pi.


````python
# Ejercicio 2
pi = 3.14159265359

def sphere_area(r):
  return round(4 * pi * (float(r)**2), 4)

def sphere_volume(r):
  return round(4/3 * pi * (float(r)**3), 4)

try:
  print("Área de esfera con radio 3= ", sphere_area(3))
  print("Volúmen de esfera con radio 3= ", sphere_volume(3))
  print("Área de esfera con radio 4= ", sphere_area(4))
  print("Volúmen de esfera con radio 4= ", sphere_volume(4))
  print("Área de esfera con radio 6= ", sphere_area(6))
  print("Volúmen de esfera con radio 6= ", sphere_volume(6))

except Exception as e:
  print(e)
  print("¡Ha habido un error al ejecutar la función! ¡¡Adiós!!")
  
````

### **Ejercicio 3**
Define dos funciones, una que convierta grados Farenheit en grados centígrados y otra que realice lo contrario, teniendo en cuenta que:
\begin{equation}
C = \frac{5}{{9}} \left(F - 32 \right)
\end{equation}


````python
# Ejercicio 3
def to_celsius(degrees):
  return round(5/9 * (float(degrees) - 32), 4)

def to_fahrenheit(degrees):
  return round((float(degrees) * 9/5) + 32, 4)

try:
  print("fahrenheit a celsius 95= ", to_celsius(95))
  print("celsius a fahrenheit 35= ", to_fahrenheit(35))
  print("fahrenheit a celsius 77= ", to_celsius(77))
  print("celsius a fahrenheit 25= ", to_fahrenheit(25))
  print("fahrenheit a celsius 14= ", to_celsius(14))
  print("celsius a fahrenheit -10= ", to_fahrenheit(-10))

except Exception as e:
  print(e)
  print("¡Ha habido un error al ejecutar la función! ¡¡Adiós!!")
````

### **Ejercicio 6**
Diseña una función que calcule el sumatorio de la diferencia entre números contiguos en una lista.

````python
# Ejercicio 6
# Ejemplo:
# sum_dif_cont([1, 3, 6, 10]) devolverá 9
# 2 + 3 + 4 (el 2 resulta de calcular 3 − 1, el 3 de calcular 6 − 3 y el 4 de calcular 10 − 6

def sum_dif_cont(numbers):
  sum = 0
  if len(numbers) > 0:
    sum = 0
    i = 1
    while i < len(numbers):
      sum += (numbers[i] - numbers[i-1])
      i += 1
    
  return sum

try:
  print(sum_dif_cont([1, 3, 6, 10]))
  print(sum_dif_cont([5, 8, 10, 12]))
  print(sum_dif_cont([12, 8, 10, 12]))

except Exception as e:
  print(e)
  print("¡Ha habido un error al ejecutar la función! ¡¡Adiós!!")
````

### **Ejercicio 7**
Diseña una función que, dada una lista de cadenas de texto, devuelva la cadena más larga. Si dos o más cadenas miden lo mismo y son las más largas, la función devolverá una cualquiera de ellas.


````python
# Ejercicio 7
# Ejemplo:
# longest_str([’Pepe’, ’Juan’, ’María’, ’Ana’]) devolverá ’María’

def longest_str(words):
  if len(words) > 0:
    sizes = {}
    for word in words:
      sizes[len(word)] = word
    
    return sizes[max(sizes.keys())]
  else:
    return "No hay coincidencias"

try:
  print(longest_str(["Pepe", "Juan", "María", "Ana"]))
  print(longest_str(["Pepe", "Juan", "María", "Ana", "Mario"]))
  print(longest_str(["Catalina","Pepe", "Juan", "María", "Ana"]))
  print(longest_str([]))

except Exception as e:
  print(e)
  print("¡Ha habido un error al ejecutar la función! ¡¡Adiós!!")
  
````

### **Ejercicio 8**
Vamos a adquirir una vivienda y para eso necesitaremos una hipoteca. La cuota mensual m que hemos de pagar para amortizar una hipoteca de h euros a lo largo de n años a un interés compuesto del i por cien anual se calcula con la fórmula:
\begin{equation}
\mathrm{m} = \frac{h r}{{1} - \left( {1} + r \right)^{-{12} n}}
\end{equation}

donde r = i/(100 * 12). Define una función que calcule la cuota (redondeada a dos decimales) dados h, n e i. Utiliza cuantas variables consideres oportuno, pero al menos r debe aparecer en la expresión cuyo valor se calcula en otra función.


````python
# Ejercicio 8
# Ejemplo: hay que pagar la cantidad de 1166,75€ al mes para amortizar una hipoteca de 150000 € en 15 años a un interés del 4.75% anual.

def monthly_interest(i):
  return i/(100 * 12)

def mortgage(h, n, i):
  r = monthly_interest(i)
  m = (h * r) / (1 - ((1 + r)**(-12*n)))
  return round(m, 2)

try:
  print("Cuota mensual:", mortgage(150000, 15, 4.75))
  print("Cuota mensual:", mortgage(150000, 10, 3.75))

except Exception as e:
  print(e)
  print("¡Ha habido un error al ejecutar la función! ¡¡Adiós!!")
````

Queremos crear un programa que trabaje con fracciones a/b. Para representar una fracción vamos a utilizar dos enteros: numerador y denominador.
Vamos a crear las siguientes funciones para trabajar con funciones:
- **leer_fracción**: La tarea de esta función es leer por teclado el numerador y el denominador. Cuando leas una fracción debes simplificarla.
- **intercambiar**: Recibe dos números como parámetros de entrada y devuelve los números ordenador de mayor a menor
- **escribir_fracción**: Esta función escribe en pantalla la fracción. Si el denominador es 1, se muestra sólo el numerador.
- **calcular_mcd**: Esta función recibe dos números y devuelve el máximo común divisor.
- **simplificar_fracción**: Esta función simplifica la fracción, para ello hay que dividir numerador y denominador por el MCD del numerador y denominador.
- **sumar_fracciones**: Función que recibe dos funciones n1/d1 y n2/d2, y calcula la suma de las dos fracciones. La suma de dos fracciones es otra fracción cuyo `numerador=n1*d2+d1*n2` y `denominador=d1*d2`. Se debe simplificar la fracción resultado.
- **restar_fracciones**: Función que resta dos fracciones: `numerador=n1*d2-d1*n2` y `denominador=d1*d2`. Se debe simplificar la
fracción resultado.
- **multiplicar_fracciones**: Función que recibe dos fracciones y calcula el
producto, para ello `numerador=n1*n2` y `denominador=d1*d2`. Se debe
simplificar la fracción resultado.
- **dividir_fracciones**: Función que recibe dos fracciones y calcula el cociente, para ello `numerador=n1*d2` y `denominador=d1*n2`. Se debe simplificar la fracción resultado.

Crear un programa que utilizando las funciones anteriores muestre el siguiente menú:
1. Sumar dos fracciones: En esta opción se piden dos fracciones y se muestra el resultado.
2. Restar dos fracciones: En esta opción se piden dos fracciones y se muestra la resta.
3. Multiplicar dos fracciones: En esta opción se piden dos fracciones y se muestra el producto.
4. Dividir dos fracciones: En esta opción se piden dos fracciones y se muestra el cociente.
5. Salir

````python
# Ejercicio 9
def read_fraction():
  a = 0
  b = 0
  while a == 0 or b == 0:
    if a == 0:
      a = int(input("Introduzca a: "))
    if b == 0:
      b = int(input("Introduzca b: "))
  
  return simplify_fraction((a, b))

def switch_fraction(f):
  a = max(f[0], f[1])
  b = min(f[0], f[1])  
  return (a, b)

def calculate_mcd(n1, n2):
  f = switch_fraction((n1,n2))
  a = f[0]
  b = f[1]
  
  remainder = a % b
  mcd = b

  while remainder != 0:
    result = int(a / b)
    mcd = remainder
    remainder = a % b

    if remainder != 0:
      a = b
      b = remainder
  return mcd

def simplify_fraction(f):
  mcd = calculate_mcd(f[0], f[1])
  return (int(f[0] / mcd), int(f[1] / mcd))

def sum_fractions(f1, f2):
  sum = ((f1[0] * f2[1] + (f2[0] * f1[1])) , (f1[1] * f2[1]))
  return simplify_fraction(sum)

def substract_fractions(f1, f2):
  subs = ((f1[0] * f2[1] - (f2[0] * f1[1])) , (f1[1] * f2[1]))
  return simplify_fraction(subs)

def multiply_fractions(f1, f2):
  mult = ((f1[0] * f2[0]) , (f1[1] * f2[1]))
  return simplify_fraction(mult)

def divide_fractions(f1, f2):
  div = ((f1[0] * f2[1]) , (f2[0] * f1[1]))
  return simplify_fraction(div)

def print_fraction(f):
  return str(f[0]) + ("/" + str(f[1]) if f[1] > 1 else "")


try:
  option = 1
  print("Trabajo con fracciones a/b")

  f1 = read_fraction()
  f2 = read_fraction()

  while option != 6:
    option = int(input("Seleccione una opción:\n1 - Sumar dos fracciones\n2 - Restar dos fracciones\n3 - Multiplicar dos fracciones\n4 - Dividir dos fracciones\n5 - Otra fracción?\n6 - Salir\nIntroduzca una opción:  "))
    
    if option == 1:
      print("La suma es:", print_fraction(sum_fractions(f1, f2)))
    elif option == 2:
      print("La resta es:", print_fraction(substract_fractions(f1, f2)))
    elif option == 3:
      print("La multiplicación es:", print_fraction(multiply_fractions(f1, f2)))
    elif option == 4:
      print("La división es:", print_fraction(divide_fractions(f1, f2)))
    elif option == 5:
      option = 1
      f1 = read_fraction()
      f2 = read_fraction()  

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de dato! ¡¡Adiós!!")
````
