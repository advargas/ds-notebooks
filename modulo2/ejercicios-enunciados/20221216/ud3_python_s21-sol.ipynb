### **Ejercicio 3**
Escribir un programa que guarde en un diccionario los precios de las siguientes frutas:
```
Plátano: 1.35
Manzana: 0.80
Pera: 0.85
Naranja: 0.70
```
pregunte al usuario por una fruta, un número de kilos y muestre por pantalla el precio de ese número de kilos de fruta. Si la fruta no está en el diccionario debe mostrar un mensaje informando de ello.


````python
# Ejercicio 3
precios = {
    "Plátano": 1.35,
    "Manzana": 0.8,
    "Pera": 0.85,
    "Naranja": 0.7
}

fruta = str(input("¿Qué fruta quiere? ")).capitalize()
if fruta not in precios.keys():
  print("No disponemos de la fruta que solicita")
else:
  try:
    kg = float(input("¿Cuántos kilogramos quiere? "))
  except:
    raise ValueError("Ha introducido un valor erróneo de kilogramos, ha de ser un valor numérico")
  
  print(kg, "kilogramos de", fruta, "salen a", kg * precios[fruta])
````

Escribir un algoritmo que pida números hasta que se introduzca un cero. Debe imprimir la suma y la media de todos los números introducidos.

````python
# Ejercicio 8
n = float(input("Introduzca un número (0, para salir): "))
nums = list()
while n:
  nums.append(n)
  n = float(input("Introduzca un número (0, para salir): "))

if not(len(nums)):
  print("No podemos realizar las operaciones si no ha introducido ningún número")
else:
  print("La suma es:", sum(nums), "\nLa media es:", sum(nums)/len(nums))
````

### **Ejercicio 9**
Escribe un programa que pida el límite inferior y superior de un intervalo. Si el límite inferior es mayor que el superior lo tiene que volver a pedir. A continuación se van introduciendo números hasta que introducimos el 0. Cuando termine el programa dará las siguientes informaciones:
- La suma de los números que están dentro del intervalo (intervalo abierto).
- Cuántos números están fuera del intervalo.
- Informar si hemos introducido algún número igual a los límites del intervalo.

````python
# Ejercicio 9
# Ejemplo:
# Introduce el límite inferior del intervalo: 10
# Introduce el límite superior del intervalo: 20
# Introduce un número (0, para salir): 5
# Introduce un número (0, para salir): 11
# Introduce un número (0, para salir): 15
# Introduce un número (0, para salir): 2
# Introduce un número (0, para salir): 22
# Introduce un número (0, para salir): 19
# Introduce un número (0, para salir): 0
# La suma de los número dentro del intervalo es 45
# La cantidad de números fuera del intervalo es 3
# No se ha introducido ningún número igual a los límites del intervalo.

try:
  inf = float(input("Introduce el límite inferior del intervalo: "))
  sup = float(input("Introduce el límite superior del intervalo: "))
  if inf > sup:
    print("El índice inferior ha de ser estrictamente menor que el superior")
    inf = float(input("Introduce el límite inferior del intervalo: "))
except:
  raise ValueError("Los valores a introducir para el límite inferior como para el límite superios han de ser numéricos")


n = float(input("Introduzca un número (0, para salir): "))
nums_int = list()
cont = 0
limits = 0
while n:
  if n > inf and n < sup:
    nums_int.append(n)
  elif not limits and (n == inf or n == sup):
    limits += 1
  else:
    cont += 1
  n = float(input("Introduzca un número (0, para salir): "))

print("La suma de los números dentro del intervalo es", sum(nums_int))
print("La cantidad de números fuera del intervalo es", cont)
if limits:
  print("Se han introducido números iguales a los límites del intervalo")
else:
  print("No se ha introducido ningún número igual a los límites del intervalo")
````

### **Ejercicio 10**
Escriba un programa que simule una hucha. El programa solicitará primero una cantidad, que será la cantidad de dinero que queremos ahorrar. A continuación, el programa solicitará una y otra vez las cantidades que se irán ahorrando, hasta que el total ahorrado iguale o supere al objetivo. El programa comprobará que las cantidades sean positivas.

````python
# Ejercicio 10
# Ejemplo:
# HUCHA
# ¿Cuántos euros quieres ahorrar?: -10
# Por favor, escribe una cantidad positiva. 
# ¿Cuántos euros quieres ahorrar?: 20.5
# ¿Cuántos euros quieres meter en la hucha?: 20.6 
# ¡Objetivo conseguido! Has ahorrado 20.6 euros.

print("HUCHA")
try:
  ahorro = float(input("¿Cuántos euros quieres ahorrar?: "))
except:
  raise ValueError("El valor introducido a de ser numérico (entero o real)")

while ahorro < 0:
  print("Por favor, escribe una cantidad positiva.")
  ahorro = float(input("¿Cuántos euros quieres ahorrar?: "))

hucha = 0
while hucha < ahorro:
  try:
    pend = float(input("¿Cuántos euros quieres meter en la hucha?: "))
    if pend < 0:
      print("Por favor, escribe una cantidad positiva.")
      continue
    else:
      hucha += pend
  except:
    raise ValueError("El valor introducido a de ser numérico (entero o real)")

print("¡Objetivo conseguido! Has ahorrado", hucha, "euros.")
````
