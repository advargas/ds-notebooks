### **Ejercicio 11**
Hacer un programa que simule los procesos de extracción de un cajero automático, permitiendo que el usuario tenga tres intentos de ingreso, en caso de error se deberá bloquear la tarjeta y saldremos de la aplicación. Si ingresa al sistema, el cajero deberá seleccionar entre la cuenta de ahorro, crédito o corriente. Una vez que haya seleccionado la cuenta, tendrá que elegir la cantidad a retirar (se deberá comprobar que tenga saldo suficiente, sino se le volverá a pedir la cantidad), se le descontará del saldo de la cuenta, se mostrará su saldo restante y nos volverá a preguntar el menú de selección de las cuentas. Con la opción 4 sale de la aplicación.


```python
# Ejercicio 1
# Ejemplo
# CAJERO AUTOMATICO
# Introduzca contraseña: 2585
# Contador de fallos en usuario y contraseña: 1 
# Introduzca contraseña: 1234
# Seleccione cuenta:
# 1 - Cuenta Ahorro
# 2 - Cuenta Crédito
# 3 - Cuenta Corriente
# 4 - Salir
# Introduzca la cuenta en la que quiere sacar: 1
# Introduzca la cantidad que desea retirar: 6000
# La cantidad seleccionada supera el saldo actual. 
# Introduzca una nueva cantidad inferior a retirar: 1000 
# Saldo restante: 4000.0
# Seleccione cuenta:
# 1 - Cuenta Ahorro
# 2 - Cuenta Crédito
# 3 - Cuenta Corriente
# 4 - Salir
# Introduzca la cuenta en la que quiere sacar: 4
print("CAJERO AUTOMATICO")
cuentas = {
    "pin": 4427,
    1: 35120, 
    2: 1000,
    3: 2360
}
cont = 0
while cont < 3:
  try:
    cand = int(input("Introduzca su contraseña: "))
    if cand != cuentas["pin"]:
      cont += 1
      print("Contador de fallos en usuario y contraseña:", cont)
    else:
      break
  except ValueError:
    raise ValueError("La contraseña ha de ser un número entero")

if cont == 3:
  print("Ha introducido erróneamente 3 veces el número PIN. "
        "\nHemos pasado a bloquear su cuenta. Contacte directamente con su banco")

else:
  ret = None
  print("Seleccione cuenta: \n1 - Cuenta Ahorro \n2 - Cuenta Crédito \n3 - Cuenta Corriente \n4 - Salir")
  try:
    acc = int(input("Introduzca la cuenta en la que quiere sacar: "))
    while acc != 4:
      if acc not in range(1, 5):
        print("Por favor, introduzca un valor entre los siguientes: 1, 2, 3 y 4")
        continue
      elif not isinstance(ret, int):
        try:
          ret = float(input("Introduzca la cantidad que desea retirar: "))
        except:
          print("Por favor, introduzca un valor numérico a retirar")

      if ret <= 0:
        print("El valor a retirar ha de ser mayor que cero")
        ret = int(input("Introduzca un valor a retirar mayor que cero: "))
      elif ret > cuentas[acc]:
        print("La cantidad seleccionada supera el saldo actual")
        ret = int(input("Introduzca una nueva cantidad inferior a retirar: "))
      else:
        cuentas[acc] -= ret
        print("Saldo restante en la cuenta " + str(acc) + ": ", cuentas[acc])
        ret = None
        print("Seleccione cuenta: \n1 - Cuenta Ahorro \n2 - Cuenta Crédito \n3 - Cuenta Corriente \n4 - Salir")
        acc = int(input("Introduzca la cuenta en la que quiere sacar: "))
  except:
    print("Debe introducir un valor entero entre los siguientes: 1, 2, 3 y 4")
```

### **Ejercicio 12**
Escribir un programa que pida al usuario un número entero positivo y muestre por pantalla todos los números impares desde 1 hasta ese número separados por comas.

```python
# Ejercicio 12
try:
  num = int(input("Proporcione un número entero positivo: "))
  if num <= 0:
    raise ValueError("El valor ha de ser positivo")
except:
  raise ValueError("El valor ha de ser un número entero")

out = list()
for n in range(1, num + 1):
  if n % 2 != 0:
    out.append(str(n))

print(", ".join(out))
```

### **Ejercicio 15**
Escribir un programa que permita al usuario introducir 6 números enteros, que pueden ser positivos o negativos. Al finalizar, mostrar la sumatoria de los números negativos y el promedio de los positivos.
No olvides que no es posible dividir por cero, por lo que es necesario evitar que el programa devuelva un error si no se introdujeron números positivos.

```python
# Ejercicio 15
t = 0
nums = list()
while t < 6:
  try: 
    n = int(input("Introduzca un valor entero negativo o positivo: "))
    if n == 0:
      print("El valor ha de ser distinto de cero")
      continue
    else:
      t += 1
      nums.append(n)
  except:
    print("El valor ha de ser un entero")

addition = 0
mean = list()
for num in nums:
  if num < 0:
    addition += num
  else:
    mean.append(num)

print("Sumatorio de los números negativos:", addition, "\nPromedio de los números positivos:", sum(mean)/len(mean))
```
