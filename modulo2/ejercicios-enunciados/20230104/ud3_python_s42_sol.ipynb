### **Ejercicio 1**

Escribir un programa que cree:
1. Un array vacío
2. Un array de ceros
3. Un array de valores 1 a 5
4. Una matriz diagonal de valores 1 a 5

Y nos muestre por cada uno de ellos sus dimensiones y el número de elementos que contiene.

```python
# Ejercicio 1
arrays = [
    np.empty((3, 5)),
    np.zeros((3, 4)),
    np.arange(1, 6),
    np.diag(list(range(1, 6)))
]

for ar in arrays:
  print("Array: \n", ar)
  print(" -- Dimensión:", ar.shape)
  print(" -- Número de elementos:", ar.size)
  print("----------------------------------------")
```

### **Ejercicio 2**
Escribir un programa que, dados dos arrays, lleve a cabo:
1. La unión
2. La intersección
3. La suma
4. La multiplicación por un valor constante


```python
# Ejercicio 2
a = np.array([1.53, 3.22])
b = np.array([7.334, 3.22])

print("Unión:", np.concatenate([a, b]))
print("Intersección:", np.intersect1d(a, b))
print("Suma:", a + b)
print("Multiplicación por una constante n = 3:", a * 3)
```

## Ejercicio 3
Escribir un programa que dados dos arrays los compare.

```python
# Ejercicio 3
a = np.array([5.445, 1.22])
b = np.array([19.56, 1.12])
print("Array a: ", a)
print("Array b: ", b)
print("a > b", np.greater(a, b))
print("a >= b", np.greater_equal(a, b))
print("a < b", np.less(a, b))
print("a <= b", np.less_equal(a, b))
```

### **Ejercicio 4**
Escribir un programa que cree un array de temperaturas en grados fahrenheit y transforme todos sus elementos a grados centígrados. Recordad que la fórmula utilizada para la conversión es:
\begin{equation}
C = \frac{5}{{9}} \left(F - 32 \right)
\end{equation}

```python
# Ejercicio 4
grados_fahrenheit = np.array([67.34, -11.11, 78.96, 1.11, 102.71, 0])
print(
    "El array de grados Fahrenheit:\n", grados_fahrenheit, 
    "\npasa a ser:\n", np.round(((5/9) * (grados_fahrenheit - 32)), 2), 
    "\nen grados Celsius.")
```
