### **Ejercicio 3**
Crear un archivo llamado personas que contendrá una lista de personas con los siguientes campos:
- id
- nombre
- apellido
- nacimiento

Leer cada línea y colocar cada empleado en un diccionario dentro de una lista. Mostrar los datos de esta. Datos dentro del fichero:
```
1;Carlos;Pérez;05/01/1989
2;Manuel;Heredia;26/12/1973 
3;Rosa;Campos;12/06/1961 
4;David;García;25/07/2006
```


````python
# Ejercicio 3
try:
  with open("personas.txt", "r") as pers:
    keys = ["id", "nombre", "apellido", "nacimiento"]
    d = dict.fromkeys(keys)
    lines = [l.strip().split(";") for l in pers.readlines()]
    for line in lines:
      for i, el in enumerate(line):
        if isinstance(d[keys[i]], list):
          d[keys[i]].append(el)
        else:
          d[keys[i]] = list([el])
except FileNotFoundError:
  print("El archivo no existe. Por favor, créalo con los datos especificados con anterioridad")
````

### **Ejercicio 4**
Diseña un programa que obtenga los 100 primeros números primos y los almacene en un fichero de texto llamado primos.txt.


````python
# Ejercicio 4
def divisores(n: int) -> list:
  d = 1
  div = list()
  while d <= n:
    if n % d == 0:
      div.append(d)
    d += 1
  return div

with open("primos.txt", "w+") as prims:
  cont = 0
  cand = 1
  while cont < 100:
    if len(divisores(cand)) <= 2:
      prims.write(str(cand) + "\n")
      cont += 1
    cand += 1
````

### **Ejercicio 6**
Escribir un programa que guarde en un fichero el ticket de compra de una venta de zapatillas. El programa deberá comenzar creando el fichero llamado “ticket” y mostrando el listado de zapatillas de la siguiente forma (con el número por delante):
1. Nike Air Max 90
2. Nike Air Force 1
3. Nike Blazer Mid 77 
4. ...

Le preguntaremos al usuario en bucle que vaya seleccionando las zapatillas que quiere comprar hasta que escriba `*`. Estas seleccionadas tendrán que guardarse en una lista que se llame “carrito”. Cuando escriba `*` tendremos que escribir las zapatillas de la lista carrito en el fichero ticket y en la ultima línea del fichero el precio total que tiene que pagar en el fichero.
```
productos = ("Nike Air Max 90", "Nike Air Force 1", "Nike Blazer Mid 77", "Air Jordan 1 Mid", "Nike SB Shane", "Nike Waffle One SE", "Nike Air Presto", "Nike React Vision")
precios = (139.99, 99.99, 109.99, 129.99, 74.99, 109.89, 85.99, 119.80) 
stocks = (43, 1, 0, 32, 10`, 12, 50, 6)
```

````python
# Ejercicio 6
# Usamos tuplas en productos y precios ya que no se van a modificar
productos = (
  "Nike Air Max 90", "Nike Air Force 1", "Nike Blazer Mid 77", "Air Jordan 1 Mid", 
  "Nike SB Shane", "Nike Waffle One SE", "Nike Air Presto", "Nike React Vision"
)
precios = (139.99, 99.99, 109.99, 129.99, 74.99, 109.89, 85.99, 119.80) 

# Usamos lista en stocks para modificarlo en función del producto solicitado
stocks = [43, 1, 0, 32, 10, 12, 50, 6]

opt = True
carrito = list()
menu = "Seleccione que producto desea:\n"
for i, zap in enumerate(productos):
  menu = menu + f"{i + 1}. {zap}\n"

while opt != "*":
  opt = str(input(menu))
  try:
    opt = int(opt)
    if opt not in range(1, len(productos) + 1):
      print(f"Por favor, introduzca un valor entero en el intervalo [1, {len(productos) + 1}]")
      continue
    else:
      if stocks[opt - 1]:
        carrito.append(productos[opt - 1])
        stocks[opt - 1] -= 1
      else:
        print("No hay disponibilidad del producto", productos[opt - 1])
        continue
  except:
    if opt == "*":
      continue
    else:
      print(f"Por favor, introduzca un valor entero en el intervalo [1, {len(productos) + 1}]")

if carrito:
  carrito.sort()
  total = 0
  with open("ticket.txt", "a+") as ticket:
    for prod in carrito:
      ticket.write(prod + "\n")
      total += precios[productos.index(prod)]
    ticket.write(f"---------------------------------------\nTotal: {total}")
  
  print("Ya puede consultar su ticket en ticket.txt")
else:
  print("No ha seleccionado ningún artículo por lo que no podemos proporcionarle un ticket")
````

### **Ejercicio 8**
Implementar un llavero de contraseñas con ficheros. Para ello nos dan un fichero, una columna donde se encuentran las aplicaciones y las contraseñas.
Tendremos que hacer:
1. Agregar nuevas contraseñas. Debes tener en cuenta que la aplicación y la contraseña que se agregue NO puede existir en el fichero. Si existe se tendrá que solicitar otra contraseña.
2. Buscar una aplicación introducida por teclado y devolver la contraseña correspondiente.
3. Deshacer la última contraseña agregada.
4. Modificar una contraseña que se ha introducido por teclado.
5. Después de realizar cada opción se preguntará si se quiere realizar otra acción. Si selecciona “S” se volverá́ a enseñar el menú́, si selecciona “N” parará el programa.

Datos dentro del fichero:
```
Gmail 20051206Correo! 
Tinder qwerasdf1234# 
Twitter Tw2022+LMP 
Tiktok Tik2022*VTP
Instagram Ins2022!MFR
```

````python
# Ejercicio 8
def agregar_contraseña(apps: list, pws: list, lines: list, path: str) -> None:
  app = str(input("Proporcione una app: ")).lower()
  while app in apps:
    app = str(input(f"Proporcione una app que no se encuentre en el llavero:\nApps existentes: {apps} - ")).lower()
  pw = str(input("Proporcione una contraseña: "))
  lines.append(f"\n{app.capitalize()} {pw}")
  with open(path, "w+") as kring_new:
    kring_new.writelines(lines)

def buscar_contraseña(apps: list, pws: list) -> None:
  app = str(input("Proporcione una app: "))
  while app not in apps:
    app = str(
        input(f"Proporcione una app que se encuentre en el llavero:\nApps existentes: {apps} - "))
  print(pws[apps.index(app.lower())])

def borrar_ultima_entrada(lines: list, path: str) -> None:
  with open(path, "w+") as kring_new:
    kring_new.writelines(lines[:-1])

def modificar_contraseña(apps: list, pws: list, lines: list, path: str):
  app = str(input("Proporcione una app: ")).lower()
  while app not in apps:
    app = str(
        input(f"Proporcione una app que se encuentre en el llavero:\nApps existentes: {apps} - ")).lower()

  pws[apps.index(app)] = str(input(f"Proporcione una nueva contraseña para {app}: "))
  to_write = list()
  for i in range(len(apps)):
    to_write.append(apps[i].capitalize() + " " + pws[i] + "\n")

  with open(path, "w+") as kring_new:
    kring_new.writelines(to_write)


opt = True
menu = "¿Qué desea hacer?:\n1. Agregar una contraseña\n2. Buscar contraseña\n3. Borrar última contraseña\n4. Modificar contraseña\n"

while opt != "N":
  opt = str(input(menu))
  try:
    opt = int(opt)
  except:
    print(f"Por favor, introduzca un valor entero entre 1, 2, 3 y 4")
    continue
  if opt not in [1, 2, 3, 4]:
    print(f"Por favor, introduzca un valor entero entre 1, 2, 3 y 4")
    continue
  else:
    try:
      with open("llavero.txt", "r") as kring:
        # Generate auxiliar lists to work with on each iteration
        apps, pws = list(), list()
        lines = [line for line in kring.readlines() if line != "\n"]
        for line in lines:
          sp_line = line.strip().split(" ")
          apps.append(sp_line[0].lower())
          pws.append(sp_line[1])
        if opt == 1:
          agregar_contraseña(apps, pws, lines, "llavero.txt")
          print("DONE!")
        elif opt == 2:
          buscar_contraseña(apps, pws)
          print("DONE!")
        elif opt == 3:
          borrar_ultima_entrada(lines, "llavero.txt")
          print("DONE!")
        else:
          modificar_contraseña(apps, pws, lines, path="llavero.txt")
          print("DONE!")
        opt = str(input("¿Desea llevar a cabo alguna otra acción? (S/N) "))
    except FileNotFoundError:
      print("La base de datos de contraseñas no existe.")
      break

print("Cerrando el llavero...")
````

## **Ejercicio 9**
En este ejercicio deberemos realizar una función que realice varias tareas sobre un fichero llamado **contador.txt** que almacenará un contador de visitas (será un número):
- Nuestro programa trabajará sobre el fichero contador.txt. Si el fichero está vacío le introduciremos el contador a 0. Si existe simplemente leeremos el valor del contador.
- Luego a partir de un argumento:
  * Si se envía el argumento inc, se incrementará el contador en uno y se mostrará por pantalla.
  * Si se envía el argumento dec, se decrementará el contador en uno y se mostrará por pantalla.
  * Si no se envía ningún argumento (o algo que no sea inc o dec), se mostrará el valor del contador por pantalla.
- Finalmente guardará de nuevo el valor del contador de nuevo en el fichero

````python
# Ejercicio 9
def contador(path: str, act: str="") -> None:
  # Initialize file if non-existent
  try:
    contxt = open(path, "r")
  except FileNotFoundError:
    with open(path, "w+") as contxt:
      contxt.write(str(0))
  
  # Open to read the counter
  with open(path, "r") as contxt_upd:
    cont = int(contxt_upd.read())

  # Action to take
  if act == "inc":
    cont += 1
    print(cont)
  elif act == "dec":
    cont -= 1
    print(cont)
  else:
    print(cont)  
  
  # Rewrite
  with open(path, "w+") as contxt_new:
    contxt_new.write(str(cont))
````
