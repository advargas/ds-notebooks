### **Ejercicio 1**
Escribir un programa que pida al usuario una palabra y muestre por pantalla el número de veces que contiene cada vocal.


````python
# Ejercicio 1
def replace_accents(vowel):
  vowel = vowel.replace('á','a')
  vowel = vowel.replace('é','e')
  vowel = vowel.replace('í','i')
  vowel = vowel.replace('ó','o')
  vowel = vowel.replace('ú','u')
  return vowel

try:
  vowels = {"a":0,"e":0,"i":0,"o":0,"u":0}
  new_word = "s"
  
  while new_word.lower() == 's':
    vowels = {"a":0,"e":0,"i":0,"o":0,"u":0}
    
    word = input("Ingresa una palabra: ")
    word = replace_accents(word.lower())
    characters = list(word)

    for char in characters:
      if char in vowels.keys():
        vowels[char] = vowels[char] + 1

    print(vowels)
    new_word = input("Ingresar otra palabra? S/N ")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 3**
Escribir un programa que guarde en un diccionario los precios de las siguientes frutas:
```
Plátano: 1.35
Manzana: 0.80
Pera: 0.85
Naranja: 0.70
```
pregunte al usuario por una fruta, un número de kilos y muestre por pantalla el precio de ese número de kilos de fruta. Si la fruta no está en el diccionario debe mostrar un mensaje informando de ello.


````python
# Ejercicio 3
try:
  fruits_price = {"Plátano": 1.35, "Manzana": 0.80, "Pera": 0.85, "Naranja": 0.70}
  new_fruit = 's'  
 
  while new_fruit.lower() == 's':
    fruit = input("Ingresa la fruta: ")

    if fruit.title() in fruits_price.keys():
      kg = int(input("Ingresa el número de kilogramos: "))
      price = round(kg * fruits_price[fruit.title()], 2)
      print("Precio total: " +  str(price) + " €")
    else:
      print("Fruta no registrada")
    
    new_fruit = input("Consultar el precio de otra fruta? S/N ")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 5**
Los tramos impositivos para la declaración de la renta en un determinado país son los siguientes:
```
Renta                      Tipo impositivo
Menos de 10000€                  5%
Entre 10000€ y 20000€           15%
Entre 20000€ y 35000€           20%
Entre 35000€ y 60000€           30%
Más de 60000€                   45%
```
Escribir un programa que pregunte al usuario su renta anual y muestre por pantalla el tipo impositivo que le corresponde.

````python
# Ejercicio 5
try:
  annual_income = float(input("Introduce tu renta anual: "))
  income_tax = 0
  
  if annual_income < 10000:
    income_tax = 5
  elif annual_income >= 10000 and annual_income < 20000:
    income_tax = 15
  elif annual_income >= 20000 and annual_income < 35000:
    income_tax = 20
  elif annual_income >= 35000 and annual_income < 60000:
    income_tax = 30
  else:
    income_tax = 45

  print("El tipo impositivo correspondiente es: " + str(income_tax) + "%")
  
except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 7**
Los alumnos de un curso se han dividido en dos grupos A y B de acuerdo al sexo y el nombre. El grupo A está formado por las mujeres con un nombre anterior a la M y los hombres con un nombre posterior a la N y el grupo B por el resto. Escribir un programa que pregunte al usuario su nombre y sexo, y muestre por pantalla el grupo que le corresponde


````python
# Ejercicio 7
try:
  new_user = "s"
  
  while new_user.lower() == 's':
    name = input("Ingresa tu nombre: ")
    gender = input("Ingresa tu género F/M: ")
    group = ''

    if (name[0].lower() < 'm' and gender.lower() == 'f') or \
       (name[0].lower() > 'n' and gender.lower() == 'm'):
      group = 'A'
    else:
      group = 'B'

    print("Tu grupo es: " + group)
    new_user = input("Ingresar otro usuario? S/N ")
  
except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 9**
Escribe un programa que pida el límite inferior y superior de un intervalo. Si el límite inferior es mayor que el superior lo tiene que volver a pedir. A continuación se van introduciendo números hasta que introducimos el 0. Cuando termine el programa dará las siguientes informaciones:
- La suma de los números que están dentro del intervalo (intervalo abierto).
- Cuántos números están fuera del intervalo.
- Informar si hemos introducido algún número igual a los límites del intervalo.

````python
# Ejercicio 9
# Ejemplo:
# Introduce el límite inferior del intervalo: 10
# Introduce el límite superior del intervalo: 20
# Introduce un número (0, para salir): 5
# Introduce un número (0, para salir): 11
# Introduce un número (0, para salir): 15
# Introduce un número (0, para salir): 2
# Introduce un número (0, para salir): 22
# Introduce un número (0, para salir): 19
# Introduce un número (0, para salir): 0
# La suma de los número dentro del intervalo es 45
# La cantidad de números fuera del intervalo es 3
# No se ha introducido ningún número igual a los límites del intervalo.
try:
  number = -1
  numbers_in = []
  numbers_out = 0
  numbers_equal = 0
  min = 0
  max = -1

  while min > max:  
    min = int(input("Introduce el límite inferior del intervalo: "))
    max = int(input("Introduce el límite superior del intervalo: "))

    if min < max:
      while number != 0:
        number = int(input("Introduce un número (0, para salir): "))
        if number != 0:
          if number == min or number == max:
            numbers_equal = numbers_equal + 1
          
          if number > min and number < max:
            numbers_in.append(number)
          else:
            numbers_out = numbers_out + 1

      print("La suma de los números dentro del intervalo es:", sum(numbers_in))
      print("La cantidad de números fuera del intervalo es:", numbers_out)

      if numbers_equal == 0:
        print("No se ha introducido ningún número igual a los límites del intervalo.")
      else:
        print("Se han introducido " + str(numbers_equal) + " número(s) igual(es) a los límites del intervalo.")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 10**
Escriba un programa que simule una hucha. El programa solicitará primero una cantidad, que será la cantidad de dinero que queremos ahorrar. A continuación, el programa solicitará una y otra vez las cantidades que se irán ahorrando, hasta que el total ahorrado iguale o supere al objetivo. El programa comprobará que las cantidades sean positivas.

````python
# Ejercicio 10
# Ejemplo:
# HUCHA
# ¿Cuántos euros quieres ahorrar?: -10
# Por favor, escribe una cantidad positiva. ¿Cuántos euros quieres ahorrar?: 20.5
# ¿Cuántos euros quieres meter en la hucha?: 20.6 
# ¡Objetivo conseguido! Has ahorrado 20.6 euros.
try:
  savings = 0
  target = -1

  while target < 0:
    target = float(input("¿Cuántos euros quieres ahorrar?: "))

    if target < 0:
      print("Por favor, escribe una cantidad positiva.")
    else:
      while savings < target:
        new_savings = float(input("¿Cuántos euros quieres meter en la hucha?: "))
        savings = savings + new_savings
    
      print("¡Objetivo conseguido! Has ahorrado " + str(round(savings, 2)) + " euros.")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````
