### **Ejercicio 11**
Hacer un programa que simule los procesos de extracción de un cajero automático, permitiendo que el usuario tenga tres intentos de ingreso, en caso de error se deberá bloquear la tarjeta y saldremos de la aplicación. Si ingresa al sistema, el cajero deberá seleccionar entre la cuenta de ahorro, crédito o corriente. Una vez que haya seleccionado la cuenta, tendrá que elegir la cantidad a retirar (se deberá comprobar que tenga saldo suficiente, sino se le volverá a pedir la cantidad), se le descontará del saldo de la cuenta, se mostrará su saldo restante y nos volverá a preguntar el menú de selección de las cuentas. Con la opción 4 sale de la aplicación.


```python
# Ejercicio 1
# Ejemplo
# CAJERO AUTOMATICO
# Introduzca contraseña: 2585
# Contador de fallos en usuario y contraseña: 1 
# Introduzca contraseña: 1234
# Seleccione cuenta:
# 1 - Cuenta Ahorro
# 2 - Cuenta Crédito
# 3 - Cuenta Corriente
# 4 - Salir
# Introduzca la cuenta en la que quiere sacar: 1
# Introduzca la cantidad que desea retirar: 6000
# La cantidad seleccionada supera el saldo actual. 
# Introduzca una nueva cantidad inferior a retirar: 1000 
# Saldo restante: 4000.0
# Seleccione cuenta:
# 1 - Cuenta Ahorro
# 2 - Cuenta Crédito
# 3 - Cuenta Corriente
# 4 - Salir
# Introduzca la cuenta en la que quiere sacar: 4
try:
  pin = 2385
  valid_pin = False
  pin_tries = 0
  account_number = 1

  accounts = {1: 5000, 2:6000, 3:10000}

  print("CAJERO AUTOMÁTICO")
  
  while not valid_pin and pin_tries < 3:
    user_pin = int(input("Introduzca contraseña: "))

    if user_pin == pin:
      valid_pin = True

      while account_number in range(1,3):
        account_number = int(input("Seleccione cuenta:\n1 - Cuenta Ahorro\n2 - Cuenta Crédito\n3 - Cuenta Corriente\n4 - Salir\nIntroduzca la cuenta de la que quiere retirar:  "))

        if account_number in range(1,3):
          non_sufficient_funds = True

          while non_sufficient_funds and accounts[account_number] > 0:
            withdraw = float(input("Introduzca la cantidad que desea retirar: "))

            if withdraw <= accounts[account_number]:
              accounts[account_number] = accounts[account_number] - withdraw
              print("Saldo restante: ", accounts[account_number])
              non_sufficient_funds = False
            else:
              print("La cantidad seleccionada supera el saldo actual.")
          else:
            if accounts[account_number] == 0:
              print("Cuenta con saldo en cero.")

    else:
      pin_tries += 1
      if pin_tries == 3:
        print("Número de intentos fallidos superado, la tarjeta será bloqueada")
      else:
        print("Contador de fallos en usuario y contraseña:", pin_tries)
    
except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
```

### **Ejercicio 12**
Escribir un programa que pida al usuario un número entero positivo y muestre por pantalla todos los números impares desde 1 hasta ese número separados por comas.

```python
# Ejercicio 12
try:
  number = 0
  odd_numbers = ''

  while number <= 0:
    number = int(input("Introduzca un número entero positivo: "))

    for i in range(1, number + 1):
      if i % 2 == 1:
        odd_numbers += str(i) + ','
    
    print(odd_numbers[:-1])

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
```

### **Ejercicio 13**
Escribir un programa que pregunte al usuario una cantidad a invertir, el interés anual y el número de años, y muestre por pantalla el capital obtenido en la inversión cada año que dura la inversión.


```python
# Ejercicio 13
try:
  p = float(input("Introduzca el monto a invertir $: "))
  r = float(input("Introduzca el interés anual %: "))
  t = int(input("Introduzca el número de años: "))
  
  for i in range(1, t + 1):
    interest = p * (r/100) * i
    print("Intereses ganados año " + str(i) + ": $" + str(interest))

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
```

### **Ejercicio 14**
Escribir un programa en el que se pregunte al usuario por una frase y una letra, y muestre por pantalla el número de veces que aparece la letra en la frase.


```python
# Ejercicio 14
import re

try:
  sentence = input("Introduzca una frase: ")
  char = input("Introduzca una letra: ")
  
  counter = len(re.findall("[" + char + char.upper() + "]+", sentence))
  print (char + " aparece " + str(counter) + " veces en '" + sentence + "'")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
```

### **Ejercicio 15**
Escribir un programa que permita al usuario introducir 6 números enteros, que pueden ser positivos o negativos. Al finalizar, mostrar la sumatoria de los números negativos y el promedio de los positivos.
No olvides que no es posible dividir por cero, por lo que es necesario evitar que el programa devuelva un error si no se introdujeron números positivos.

```python
# Ejercicio 15
try:
  positives = []
  negatives = []

  for i in range(1, 7):
    number = int(input("Introduzca un número entero " + str(i) + ": "))
    
    if number >= 0:
      positives.append(number)
    else:
      negatives.append(number)

  print("Suma de negativos: ", sum(negatives))
  print("Promedio de positivos: ", round(sum(positives) / len(positives), 2))

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
```
