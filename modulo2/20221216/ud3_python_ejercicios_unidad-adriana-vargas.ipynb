Escribe un programa que almacene las asignaturas de un curso (por ejemplo Matemáticas, Física, Química, Historia y Lengua) en una lista, pregunte al usuario la nota que ha sacado en cada asignatura, y después las muestre por pantalla con el mensaje:
```
En {asignatura} has sacado {nota}
```
Donde {asignatura} es cada una de las asignaturas de la lista y {nota} cada una de las correspondientes notas introducidas por el usuario.

````python
# Ejercicio 9
try:
  subjects = ["Matemáticas","Física","Química","Historia","Lengua"]
  grades = {}

  for subject in subjects:
    grade = float(input("Ingresa tu nota para " + subject + ": "))
    grades[subject] = grade

  for subject in grades.keys():
    print("En", subject, "has sacado", grades[subject])

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 10**
Escribe un programa que pregunte al usuario los números ganadores de la lotería primitiva, los almacene en una lista y los muestre por pantalla ordenados de menor a mayor.

````python
# Ejercicio 10
try:
  count = 1
  numbers = []

  while count <= 6:
    number = int(input("Ingresa un número para la lotería primitiva: "))

    if number >= 1 and number <= 49 and number not in numbers:
      numbers.append(number)
      count = count + 1
    else:
      print ("Número repetido o fuera de rango 1-49")

  numbers.sort()
  print("Los números seleccionados son", numbers)

except:
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 11**
Escribe un programa que almacene las asignaturas de un curso (por ejemplo Matemáticas, Física, Química, Historia y Lengua) en una lista, pregunte al usuario la nota que ha sacado en cada asignatura y elimine de la lista las asignaturas aprobadas. Al final el programa debe mostrar por pantalla las asignaturas suspensas.

````python
# Ejercicio 11
try:
  subjects = ["Matemáticas","Física","Química","Historia","Lengua"]
  grades = {}
  passed_grades = []

  for subject in subjects:
    grade = float(input("Ingresa tu nota para " + subject + ": "))
    grades[subject] = grade

  for subject in grades.keys():
    if grades[subject] >= 3.0:
      passed_grades.append(subject)

  for subject in passed_grades:
    del grades[subject]
  
  print("Asignaturas Suspensas:")
  print(grades)

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 12**
Escribe un programa que pregunte al usuario sus datos personales (nombre, edad, dirección y teléfono), y lo guarde en un diccionario. Después debe mostrar por pantalla el mensaje:
```
{nombre} tiene {edad} años, vive en {dirección} y su número de teléfono es {teléfono}
```

````python
# Ejercicio 12
try:
  user_data = {}
  user_keys = ["Edad", "Dirección", "Teléfono"]
  new_user = "s"

  while new_user.lower() == 's':
    name = input("Ingresa el nombre: ")
    user_data[name] = []

    for data in user_keys:
      info = input("Ingresa " + data + ": ")
      user_data[name].append(info)
    
    new_user = input("Ingresar otro usuario? S/N ")
    
  for name in user_data.keys():
    print(name + " tiene " + user_data[name][0] + " años, vive en " + user_data[name][1] + " y su número de teléfono es " + user_data[name][2])

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````

### **Ejercicio 13**
Escribe un programa que cree un diccionario de traducción español-inglés. El usuario introducirá las palabras en español e inglés separadas por dos puntos, y cada par `{palabra}:{traducción}` separados por comas. El programa debe crear un diccionario con las palabras y sus traducciones. Después pedirá una frase en español y utilizará el diccionario para traducirla palabra a palabra. Si una palabra no está en el diccionario debe dejarla sin traducir.


````python
# Ejercicio 13
try:
  translator = {}
  new_translation = "s"
  
  en_dictionary = input("Ingresa el diccionario español/inglés así -> {palabra1}:{traducción1},{palabra2}:{traducción2}\n")

  for translation in en_dictionary.split(','):
    es_word = translation.split(':')[0]
    en_word = translation.split(':')[1]
    translator[es_word.lower()] = en_word.lower()
  
  while new_translation.lower() == 's':
    sentence = input("Ingresa la frase a traducir: ")
    
    translated_sentence = ''
    for word in sentence.split(' '):
      if word.lower() in translator:
        translated_sentence = translated_sentence + " " + translator[word.lower()]
      else:
        translated_sentence = translated_sentence + " " + word.lower()
    
    print("Frase en inglés: " + translated_sentence.capitalize())
    new_translation = input("Ingresar otra frase a traducir? S/N ")

except Exception as e:
  print(e)
  print("¡Ha habido un error al convertir el tipo de datos! ¡¡Adiós!!")
````
