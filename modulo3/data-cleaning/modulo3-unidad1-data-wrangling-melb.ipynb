```python
# Antes de convertir 
melb.Date[:2]
melb['Date'] = pd.to_datetime(melb['Date'])
# Después de convertir
melb.Date[:2]
```

## Cambio de tipos de datos

```python
#Antes de convertir 
melb['Propertycount'][:2]
melb['Propertycount'] = melb['Propertycount'].astype('int')
# Después de convertir
melb['Propertycount'][:2]
```

## Sustitución de valores

```python
#Antes de convertir 
melb.Type.unique()
melb.Type.replace({'h': 'house', 'u': 'unit', 't': 'town_house'}, inplace=True)
# Después de convertir
melb.Type.unique()
melb.head()
```

## Tipo de datos de categoría

```python
melb.Type.dtypes
melb.Type.memory_usage()
```

```python
melb['Type'] = melb['Type'].astype('category')
melb['Type'].memory_usage()
```
